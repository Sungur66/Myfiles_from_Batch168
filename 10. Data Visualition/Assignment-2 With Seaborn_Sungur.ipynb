## 2: Plot 12 most dangerous cities

```python
pku['danger'] = pku.groupby('city')['id'].transform('count')

top_cities = pku[['city', 'danger']].drop_duplicates().sort_values(by='danger', ascending=False).head(12)

print(top_cities)
```

```python
olay_sayisi = np.array(list(top_cities["danger"]))
olay_sayisi
```

```python
olay_sehir = np.array(list(top_cities["city"]))
olay_sehir
```

```python
# with seaborn
```

```python
sns.barplot(x=olay_sehir, y=olay_sayisi)

plt.title("Most Dangerous Cities", color="r")

plt.xticks(rotation = 45, color="r") ; # bununla sanirim yazilari egebiliriz
```

## 3: Plot 20 most dangerous states

```python
pku['dang_state'] = pku.groupby('state')['id'].transform('count')

top_states = pku[['state', 'dang_state']].drop_duplicates().sort_values(by='dang_state', ascending=False).head(20)

print(top_states)
```

```python
# with seaborn barplot
```

```python
olay_sayi_state = np.array(list(top_states["dang_state"]))
olay_sayi_state
```

```python
olay_state = np.array(list(top_states["state"]))
olay_state
```

```python
sns.barplot(x=olay_state, y=olay_sayi_state)

plt.title("Most Dangerous States", color="c");
```

```python
# with seaborn countplot
```

```python
sns.barplot(x=olay_state, y=olay_sayi_state)

plt.xlabel("state")
plt.ylabel("count")
plt.title("Most Dangerous States", color="c");
```
