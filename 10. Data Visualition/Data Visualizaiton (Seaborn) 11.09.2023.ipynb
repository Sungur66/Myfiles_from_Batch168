# Scatter Plot

```python
%matplotlib notebook
sns.scatterplot(x = "total_bill", y = "tip", data = tips);
```

```python
%matplotlib notebook
plt.figure(figsize = (9,5))
sns.regplot(data = tips, x = "total_bill", y = "tip");  
# gÃ¶lge gÃ¼ven araligi ni belirtiyor. gÃ¶lge genisliyorsa gÃ¼ven azaliyor. 
# regresyon = bagimsiz degiskenin, bagimli degiskenle aciklama, aciklamaya calsima y=a + bx1 + cx2 + e
# regresyon cizcigisi best line, cizginin altindaki ve Ã¼stÃ¼ndeki degerlerin tolami sifir olur.
```

# Hue

```python
%matplotlib notebook
plt.figure(figsize = (12,7))
sns.scatterplot(y = "tip", data = tips, hue="day", x = "total_bill");
```

```python
%matplotlib notebook
plt.figure(figsize = (12,7))
sns.scatterplot(y = "tip", data = tips, hue="sex", x = "total_bill");
```

```python
%matplotlib notebook
plt.figure(figsize = (12,7))
sns.scatterplot(y = "tip", data = tips, hue="smoker", x = "total_bill");
```

# Style

```python
%matplotlib notebook
plt.figure(figsize = (12,7))
sns.scatterplot(y = "tip", data = tips, x = "total_bill", style = "day");
```
