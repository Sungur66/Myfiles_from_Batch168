## Pie Chart

- Data seti oluşturma

```python
x = ['Kategori 1', 'Kategori 2', 'Kategori 3', 'Kategori 4']
```

- Grafik oluşturma

```python
import matplotlib.pyplot as plt

# Veri
sizes = [15, 30, 45, 10]

# Pasta grafiğini oluştur
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=x)     # kategorik, az sınıflı (5 deger), oransal değer grafiği

# Göster
plt.axis('equal')
plt.show()
```

**Kod Açıklaması**

Bu grafikte:

- labels ve sizes, pie chart'ın etiketleri ve boyutlarıdır.

- colors, farklı dilimler için renkleri belirtir.

- explode, belirli bir dilimi (bu örnekte ilk dilimi) dışarı çıkarır. Bu, belirli bir dilimi vurgulamak için kullanılır.

- autopct, her bir dilimin yüzde kaç olduğunu gösterir.

- shadow=True, dilimlere gölge ekler.

- startangle=90, ilk dilimin nereden başlayacağını belirtir.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Seaborn stilini ayarla
#plt.style.use('classic')
sns.set(style="whitegrid")

# Tips veri setini yükle
tips = sns.load_dataset("tips")

# Yemek yenen günlerin frekansını hesapla
day_counts = tips['day'].value_counts()

# Pie chart için parametreler
labels = day_counts.index
sizes = day_counts.values
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice (yani ilk dilimi dışarı çıkar)

# Pie chart oluştur
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)

# Eksenleri eşit yap
plt.axis('equal')

# Başlık ekleyin
plt.title('Yemek Yenen Günlerin Oransal Dağılımı')

# Göster
plt.show()
```

**NOT**

Size hesaplanırken;

Pie chart'ta her bir dilimin "büyüklüğü" (yani açısal boyutu), o dilimin veri setindeki oransal frekansına bağlıdır. Yani, bir dilimin büyüklüğü, o dilimin toplamda kaç kez görüldüğüne göre hesaplanır.

Örneğin, day_counts değişkeni şu şekilde bir frekans dağılımı içeriyorsa:

Cumartesi (Sat): 60
Pazar (Sun): 50
Perşembe (Thur): 30
Cuma (Fri): 10
Toplamda 150 yemek (60 + 50 + 30 + 10) olduğunu görebiliriz.

Bu durumda, her bir günün pie chart'taki dilim büyüklüğü şu şekilde hesaplanır:



Bu hesaplamalar, Matplotlib'in pie fonksiyonu tarafından otomatik olarak yapılır. sizes değişkenine atadığınız değerler (yani frekanslar), bu hesaplamalar için kullanılır.

# Seaborn

## Distributions Plots (Dağılım Grafikleri)

### Displot

displot fonksiyonu, Seaborn kütüphanesinde bir değişkenin dağılımını görselleştirmek için kullanılır. Bu fonksiyon, histogram, KDE plot ve rug plot gibi farklı türde görselleştirmeler oluşturabilir. displot fonksiyonu, bir değişkenin hangi değerlerde daha yoğun olduğunu, hangi değerlerde daha seyrek olduğunu gösterir.

Örneğin, tips veri setinde toplam hesap (total_bill) miktarlarının dağılımını görmek için displot kullanabiliriz.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Tips veri setini yükle
tips = sns.load_dataset("tips")

# Displot oluşturma
sns.displot(tips['total_bill'])

# Eksen etiketleri ve başlık
plt.xlabel('Toplam Hesap Miktarı ($)')
plt.ylabel('Frekans')
plt.title('Toplam Hesap Miktarlarının Dağılımı')

# Grafiki göster
plt.show()
```

**Output Açıklaması**

Bu grafikte:

- X ekseninde toplam hesap miktarları yer alır.

- Y ekseninde ise bu miktarların frekansı (sıklığı) gösterilir.

- Displot da, toplam hesap miktarlarının sıklığını gösterir. Yüksek sütunlar, o miktarların daha sık görüldüğü yerlerdir.

displot, bir değişkenin dağılımını anlamak için oldukça yararlı ve esnek bir fonksiyondur. Farklı parametreler (örneğin, kde, rug) ile görselleştirmeyi farklı şekillerde özelleştirebiliriz. Örneğin, bu grafik sayesinde çoğu toplam hesap miktarının 10-20 dolar arasında olduğunu görebiliriz.

### Histplot

histplot fonksiyonu, Seaborn kütüphanesinde bir değişkenin histogramını çizmek için kullanılır. Histogram, bir değişkenin farklı aralıklardaki (bin) frekansını gösterir. Yani, bir değişkenin hangi değerlerde daha yoğun olduğunu, hangi değerlerde daha seyrek olduğunu gösterir.

Örneğin, tips veri setinde toplam hesap (total_bill) miktarlarının dağılımını görmek için histplot kullanabiliriz.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Tips veri setini yükle
tips = sns.load_dataset("tips")

# Histplot oluşturma
sns.histplot(tips['total_bill'], kde=True)

# Eksen etiketleri ve başlık
plt.xlabel('Toplam Hesap Miktarı ($)')
plt.ylabel('Frekans')
plt.title('Toplam Hesap Miktarlarının Dağılımı')

# Grafiki göster
plt.show()
```

**Output Açıklaması**

Bu grafikte:

- X ekseninde toplam hesap miktarları yer alır.

- Y ekseninde ise bu miktarların frekansı (sıklığı) gösterilir.

- Her bir sütun (bin), belirli bir aralıktaki toplam hesap miktarlarının frekansını gösterir. Örneğin, 10-20 dolar arasındaki toplam hesap miktarlarının frekansı en yüksektir.

- KDE çizgisi (mavi), toplam hesap miktarlarının yoğunluğunu gösterir ve histogramın pürüzsüz bir versiyonunu sunar.

histplot, bir değişkenin dağılımını anlamak için oldukça yararlıdır. Farklı bins parametreleri ile görselleştirmeyi farklı şekillerde özelleştirebiliriz. Örneğin, bu grafik sayesinde çoğu toplam hesap miktarının 10-20 dolar arasında olduğunu görebiliriz.

### KDE Plot

KDE (Kernel Density Estimation) plot, bir veri setindeki değişkenin olasılık yoğunluk fonksiyonunu görselleştirmek için kullanılır. Yani, bir değişkenin hangi değerlerde daha yoğun olduğunu, hangi değerlerde daha seyrek olduğunu gösterir. Seaborn kütüphanesinde kdeplot fonksiyonu ile KDE grafikleri oluşturabiliriz.

Örneğin, tips veri setinde bahşiş (tip) miktarlarının dağılımını görmek için kdeplot kullanabiliriz.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Tips veri setini yükle
tips = sns.load_dataset("tips")

# KDE plot oluşturma
sns.kdeplot(tips['tip'])

# Eksen etiketleri ve başlık
plt.xlabel('Bahşiş Miktarı ($)')
plt.ylabel('Yoğunluk')
plt.title('Bahşiş Miktarlarının Dağılımı')

# Grafiki göster
plt.show()
```

**Output Açıklaması**

Bu grafikte:

- X ekseninde bahşiş miktarları yer alır.

- Y ekseninde ise bu bahşiş miktarlarının yoğunluğu gösterilir.

- Grafikteki çizgi, bahşiş miktarlarının yoğunluğunu gösterir. Örneğin, çizginin yüksek olduğu yerler, o bahşiş miktarlarının daha sık görüldüğü yerlerdir.

KDE plot, bir değişkenin dağılımını anlamak için oldukça yararlıdır. Histogramlara benzer bir bilgi sağlar, ancak daha pürüzsüz bir görselleştirme sunar. Örneğin, bu grafik sayesinde çoğu bahşiş miktarının 2-4 dolar arasında olduğunu görebiliriz.

### Rug Plot

rugplot, bir değişkenin dağılımını göstermek için kullanılan çok basit bir görselleştirme türüdür. Bu tür bir grafik, değişkenin aldığı her bir değer için X ekseninde küçük bir çizgi çizer. rugplot, veri setindeki bireysel gözlemlerin nerede yoğunlaştığını görmek için yararlıdır.

Örneğin, tips veri setinde bahşiş (tip) miktarlarının dağılımını görmek için rugplot kullanabiliriz.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Tips veri setini yükle
tips = sns.load_dataset("tips")

# Rugplot oluşturma
sns.rugplot(x="tip", data = tips, height = 0.5)

# Eksen etiketleri ve başlık
plt.xlabel('Bahşiş Miktarı ($)')
plt.title('Bahşiş Miktarlarının Dağılımı')

# Grafiki göster
plt.show()
```

**Output Açıklaması**

Bu grafikte:

- X ekseninde bahşiş miktarları yer alır.

- Her bir küçük çizgi, bir gözlemin bahşiş miktarını temsil eder.

- Çizgilerin yoğun olduğu bölgeler, o bahşiş miktarlarının daha sık görüldüğü yerlerdir.

rugplot, bir değişkenin dağılımını anlamak için oldukça basit ve etkili bir yoldur. Ancak, çok fazla gözlem olduğunda, çizgiler birbirine karışabilir ve görselleştirme karmaşık hale gelebilir. Bu tür durumlar için genellikle KDE plot veya histogram gibi diğer görselleştirme türleri tercih edilir. Ancak rugplot, veri setindeki bireysel gözlemlerin nerede yoğunlaştığını hızlı bir şekilde görmek için yararlıdır.

## Categorical Plots (Kategorik Grafikler)

## Bar plot

- barplot kategorik verileri görselleştirmek için kullanılır. Ancak countplot'tan farklı olarak, barplot bir kategorinin ortalamasını veya başka bir istatistiksel ölçümünü gösterir.

- Örneğin, tips veri setinde hangi günler daha yüksek bahşiş verildiğini görmek isteyebiliriz. Bunun için day sütununu kategorik değişken olarak, tip sütununu ise gözlemlemek istediğimiz sayısal değişken olarak kullanabiliriz.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Tips veri setini yükle
tips = sns.load_dataset("tips")

# Barplot oluşturma
sns.barplot(x='day', y='tip', data=tips)

# Eksen etiketleri ve başlık
plt.xlabel('Gün')
plt.ylabel('Ortalama Bahşiş ($)')
plt.title('Hangi Günler Daha Yüksek Bahşiş Veriliyor?')

# Grafiki göster
plt.show()
```

Bu grafikte:

- X Ekseni
X ekseninde yine haftanın günleri yer alır: Perşembe, Cuma, Cumartesi ve Pazar.

- Y Ekseni
Y ekseninde ise her bir gün için verilen bahşişlerin ortalaması yer alır.

Bu grafik sayesinde, hangi günler daha yüksek bahşiş verildiğini kolayca görebiliriz. Örneğin, eğer Cumartesi günü için yüksek bir sütun varsa, bu restorana en çok Cumartesi günü yüksek bahşiş verildiği anlamına gelir.

Ayrıca barların üstündeki çizgiler ise;

- "güven aralığı" olarak bilinir. Bu çizgiler, verinin ne kadar güvenilir olduğu hakkında bir fikir verir. Eğer bir barın güven aralığı çok büyükse, bu o gün için bahşiş miktarının çok değişken olduğunu gösterir. Eğer güven aralığı çok küçükse, bu bahşiş miktarının genellikle aynı olduğunu gösterir.

- Basit bir örnek vermek gerekirse, diyelim ki bir sınıfta bir sınav yapıldı. Eğer herkes benzer puanlar aldıysa, sınıfın ortalaması çok "güvenilir" olacaktır. Ama eğer puanlar çok farklıysa (birkaç kişi çok yüksek, birkaç kişi çok düşük aldıysa), ortalamayı tek başına kullanmak yanıltıcı olabilir. İşte bu yüzden, ortalamayı bir "güven aralığı" ile birlikte göstermek daha bilgilendirici olabilir.

- Bu çizgiler, verinin ne kadar dalgalı olduğu veya ne kadar "güvenilir" olduğu hakkında bilgi verir. Eğer bir bar grafiğinde bu çizgileri görmüyorsanız, genellikle bu, verinin oldukça tutarlı olduğu anlamına gelir. Ancak, bu çizgiler çok uzunsa, bu, o kategorideki değerlerin oldukça farklı olduğu ve ortalamayı yorumlarken dikkatli olunması gerektiği anlamına gelir.

**Kutu (Box) ve Çizgiler**

- Kutunun üst ve alt kenarları, bahşişlerin "çeyrekliklerini" (Q1 ve Q3) gösterir.

- Kutunun dışındaki çizgiler ("bıyıklar"), normal olarak kabul edilen bahşiş miktarlarının en yüksek ve en düşük değerlerini gösterir.

- Kutu dışında kalan noktalar, "aykırı değerler" olarak kabul edilir. Yani, bu değerler normalden çok daha yüksek veya düşük bahşişlerdir.

Bu grafik sayesinde, hangi günlerde ne kadar bahşiş verildiği, bahşiş miktarlarının ne kadar değişken olduğu ve aykırı değerlerin olup olmadığı gibi birçok şeyi anlayabiliriz.

### swarmplot

Swarm plot, bir "beehive" (arı kovanı) grafiği olarak da bilinir ve kategorik veri için bir dağılım grafiği çizer. Bu grafik türü, her bir veri noktasını kategorik bir eksen boyunca gösterir, böylece veri noktaları birbirinin üzerine binmez. Bu, veri setindeki her bir noktanın görsel olarak temsil edildiği anlamına gelir. Swarm plot genellikle kutu (box) veya violin grafikleri ile birlikte kullanılır, böylece veri dağılımının yanı sıra bireysel veri noktaları da görülebilir.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Tips veri setini yükle
tips = sns.load_dataset("tips")

# Swarm plot oluşturma
sns.swarmplot(x="day", y="tip", data=tips)

# Başlık ve etiketler
plt.title('Farklı Günlerde Alınan Bahşiş Miktarları (Swarm Plot)')
plt.xlabel('Gün')
plt.ylabel('Bahşiş ($)')

# Grafiki göster
plt.show()
```

**Output Açıklaması**

Bu Grafikte
- Her bir nokta, bir müşterinin bıraktığı bahşişi temsil eder.

- Noktaların yatay konumu, bahşişin alındığı günü (Perşembe, Cuma, Cumartesi, Pazar) gösterir.

- Noktaların dikey konumu, bahşiş miktarını gösterir.

Bu grafik sayesinde, farklı günlerde alınan bahşiş miktarlarını ve dağılımlarını detaylı bir şekilde gözlemleyebiliriz. Örneğin, Cumartesi günü daha fazla sayıda bahşiş alındığını, fakat bu bahşişlerin miktarlarının farklılık gösterdiğini anlayabiliriz.

### Violin Plot

Violin plot, bir veri dağılımının yoğunluğunu ve olasılık yoğunluğunu göstermek için kullanılır. Temel olarak bir kutu grafiği (box plot) ile bir yoğunluk grafiğinin (kernel density plot) birleşimidir. Violin plot, verinin nasıl dağıldığını, medyan ve çeyreklik değerlerini, aynı zamanda veri yoğunluğunu da gösterir. Özellikle birden fazla kategorik değişkenin sayısal bir değişkenle ilişkisini görmek için çok yararlıdır.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Tips veri setini yükle
tips = sns.load_dataset("tips")

# Violin plot oluşturma
sns.violinplot(x="day", y="tip", data=tips)

# Başlık ve etiketler
plt.title('Farklı Günlerde Alınan Bahşiş Miktarları')
plt.xlabel('Gün')
plt.ylabel('Bahşiş ($)')

# Grafiki göster
plt.show()
```

Bu Grafikte
- Her bir "violin" farklı bir günü temsil eder (Perşembe, Cuma, Cumartesi, Pazar).

- Violin'lerin genişliği, o gün alınan bahşiş miktarlarının yoğunluğunu gösterir. Geniş alanlar, daha yoğun veri noktalarını temsil eder.

- Beyaz nokta, her gün için medyan bahşiş miktarını gösterir.

Bu şekilde, farklı günlerde alınan bahşiş miktarlarını ve dağılımlarını kolayca karşılaştırabiliriz.

```python
#groupby('day') ile veriyi günlerine göre grupluyoruz. 
#Ardından, ['tip'].sum() ile her bir gün için toplam bahşiş miktarını hesaplıyoruz.

# Gün başına toplam bahşiş miktarını hesapla
total_tips_per_day = tips.groupby('day')['tip'].sum()

#Yine, groupby('day') ile veriyi günlerine göre grupluyoruz. 
#Bu kez, ['tip'].mean() ile her bir gün için ortalama bahşiş miktarını hesaplıyoruz.

# Gün başına ortalama bahşiş miktarını hesapla
average_tips_per_day = tips.groupby('day')['tip'].mean()

#idxmax() fonksiyonu, en yüksek toplam bahşişi gösteren günün adını döndürür.

# En fazla toplam bahşiş alınan gün
max_total_tips_day = total_tips_per_day.idxmax()

#idxmax() fonksiyonu, en yüksek ortalama bahşişi gösteren günün adını döndürür.

# En yüksek ortalama bahşiş alınan gün
max_average_tips_day = average_tips_per_day.idxmax()
#Not: idxmax() fonksiyonu, pandas serisindeki en büyük değere sahip olan indeksi 
#(veya etiketi) döndürür. Eğer seride birden fazla en büyük değer varsa, 
#ilk karşılaşılanın indeksini döndürür.

#En fazla toplam ve ortalama bahşiş alınan günleri, 
#yanı sıra bu günlerde alınan bahşiş miktarlarını ekrana yazdırıyoruz.

print(f"En fazla toplam bahşiş alınan gün: {max_total_tips_day} (Toplam Bahşiş: {total_tips_per_day[max_total_tips_day]:.2f})")
print(f"En yüksek ortalama bahşiş alınan gün: {max_average_tips_day} (Ortalama Bahşiş: {average_tips_per_day[max_average_tips_day]:.2f})")

#Bu bölümde, total_tips_per_day değişkenini ekrana yazdırıyoruz. 
#Bu değişken, her bir gün için toplam bahşiş miktarını içerir. 
#Örneğin, Perşembe günü toplamda ne kadar bahşiş alındı, Cuma günü ne kadar, gibi.

# Günlerin toplam ve ortalama bahşiş miktarlarını yazdır
print("\nGünlerin Toplam Bahşiş Miktarları:")
print(total_tips_per_day)

#Bu bölümde, average_tips_per_day değişkenini ekrana yazdırıyoruz. 
#Bu değişken, her bir gün için ortalama bahşiş miktarını içerir. 
#Yani, Perşembe günü ortalama olarak ne kadar bahşiş alındı, 
#Cuma günü ne kadar, gibi.

print("\nGünlerin Ortalama Bahşiş Miktarları:")
print(average_tips_per_day)
```

## Comparison Plots (Karşılaştırma Grafikleri)

### Joint Plot

jointplot da Seaborn kütüphanesinde yer alır ve iki sayısal değişken arasındaki ilişkiyi görselleştirmek için kullanılır. Bu grafik türü, bir scatter plot ile iki değişkenin histogramlarını bir arada gösterir.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Tips veri setini yükle
tips = sns.load_dataset("tips")

# Jointplot oluşturma
sns.jointplot(x='total_bill', y='tip', data=tips)

# Grafiki göster
plt.show()
```

**Output Açıklaması**

Bu grafikte:

- Merkezi Scatter Plot
Merkezdeki scatter plot, total_bill ve tip değişkenleri arasındaki ilişkiyi gösterir. Yani, toplam hesap miktarı arttıkça bahşiş miktarının da artıp artmadığını bu grafik sayesinde görebiliriz.

- Kenar Histogramlar
Grafikteki üst ve sağ kenarlarda, total_bill ve tip için ayrı ayrı histogramlar yer alır. Bu histogramlar, her bir değişkenin kendi içinde nasıl bir dağılım gösterdiğini anlatır.

- jointplot, iki değişken arasındaki ilişkiyi anlamak ve aynı zamanda bu değişkenlerin kendi dağılımlarını görmek için oldukça yararlıdır. Örneğin, bu grafik sayesinde toplam hesap miktarı arttıkça genellikle bahşiş miktarının da arttığını gözlemleyebiliriz.

### Pair Plot

pairplot Seaborn kütüphanesinde yer alır ve bir veri setindeki sayısal değişkenler arasındaki ilişkileri görselleştirmek için kullanılır. Yani, bir veri setindeki tüm sayısal değişkenlerin birbiriyle nasıl bir ilişkisi olduğunu hızlı bir şekilde görmemizi sağlar.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Tips veri setini yükle
tips = sns.load_dataset("tips")

# Pairplot oluşturma
sns.pairplot(tips, hue = "day")

# Grafiki göster
plt.show()
```

**Output Açıklaması**

Bu grafikte:

- Her bir kare, iki farklı değişken arasındaki ilişkiyi gösterir. Örneğin, ilk satırın ikinci sütununda total_bill ve tip arasındaki ilişki gösterilir.
- Eğer bir değişken kendisiyle karşılaştırılıyorsa (örneğin, total_bill ve total_bill), o zaman bu karede değişkenin dağılımı (histogram) gösterilir.
- pairplot, bir veri setindeki sayısal değişkenler arasındaki ilişkileri hızlı bir şekilde görmek için çok yararlıdır. Örneğin, toplam hesap miktarı arttıkça bahşiş miktarının da artıp artmadığını bu grafik sayesinde görebiliriz.

### Catplot

catplot Seaborn kütüphanesinde yer alır ve kategorik verileri görselleştirmek için çok yönlü bir fonksiyondur. catplot, farklı türdeki kategorik görselleştirmeleri (örneğin, bar plots, violin plots, box plots) kolayca oluşturmanıza olanak tanır.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Tips veri setini yükle
tips = sns.load_dataset("tips")

# Catplot oluşturma (bar plot türünde)
sns.catplot(x='day', y='tip', data=tips) #kind="bar", kind="box", kind = "violin"

# Eksen etiketleri ve başlık
plt.xlabel('Gün')
plt.ylabel('Ortalama Bahşiş ($)')
plt.title('Hangi Günler Daha Yüksek Bahşiş Veriliyor?')

# Grafiki göster
plt.show()
```

**Output Açıklaması**

Bu grafikte:

- X Ekseni
X ekseninde haftanın günleri yer alır: Perşembe, Cuma, Cumartesi ve Pazar.

- Y Ekseni
Y ekseninde ise her bir gün için verilen bahşişlerin ortalaması yer alır.

- kind parametresi ile farklı türde grafikler oluşturabiliriz. Örneğin, kind='box' yaparsak, bir box plot oluştururuz. kind='violin' yaparsak, bir violin plot oluştururuz.

- catplot, kategorik verileri görselleştirmek için oldukça esnek ve yararlıdır. Farklı kind parametreleri ile veriyi farklı açılardan inceleyebiliriz.

### Matrixplot

Korelasyon matrisi genellikle değişkenler arasındaki ilişkiyi anlamak için kullanılır. Korelasyon katsayıları -1 ile 1 arasında değerler alır. 1'e yakın bir korelasyon pozitif bir ilişkiyi, -1'e yakın bir korelasyon negatif bir ilişkiyi ve 0 ise değişkenler arasında bir ilişki olmadığını gösterir.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Tips veri setini yükle
tips = sns.load_dataset("tips")

# Korelasyon matrisini hesapla
correlation_matrix = tips.corr()

# Isı haritası oluşturma
sns.heatmap(correlation_matrix, annot=True)

# Eksen etiketleri ve başlık
plt.title('Korelasyon Matrisi')

# Grafiki göster
plt.show()
```

**Output Açıklaması**

Bu grafikte:

- Her bir hücre, iki değişken arasındaki korelasyon katsayısını gösterir. Örneğin, total_bill ve tip arasındaki korelasyon 0.68'dir, bu yüksek bir pozitif korelasyonu işaret eder.

- Diagonal (köşegen) üzerindeki hücreler, bir değişkenin kendisiyle olan korelasyonunu gösterir, bu yüzden değerleri 1'dir.

- Renkler, korelasyonun yönünü ve şiddetini gösterir. Mavi renkler pozitif, kırmızı renkler negatif korelasyonu gösterir. Örneğin, total_bill ve tip arasında mavi bir renk var, bu da aralarında pozitif bir ilişki olduğunu gösterir.

Bu görselleştirme, değişkenler arasındaki ilişkileri hızlı bir şekilde anlamamıza yardımcı olur. Örneğin, total_bill ve tip arasında pozitif bir korelasyon olduğunu görebiliriz, bu da toplam hesap miktarı arttıkça bahşiş miktarının da genellikle arttığını gösterir.

### Grid Plots

#### pairgrid

PairGrid fonksiyonu Seaborn kütüphanesinde yer alır ve bir veri setindeki sayısal değişkenler arasındaki ilişkileri görselleştirmek için kullanılır. Ancak pairplot'tan farklı olarak, PairGrid daha esnektir ve her bir alt grafiğe farklı türde grafikler eklememize olanak tanır.

Örneğin, tips veri setinde toplam hesap (total_bill), bahşiş (tip) ve yemek yiyen kişi sayısı (size) gibi farklı sayısal değişkenler var. Bu değişkenler arasındaki ilişkileri görmek için PairGrid kullanabiliriz.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Tips veri setini yükle
tips = sns.load_dataset("tips")

# PairGrid oluşturma
g = sns.PairGrid(tips)

# Üst üçgen matrise scatter plot ekleyelim
g.map_upper(sns.scatterplot)

# Köşegen matrise histogram ekleyelim
g.map_diag(sns.histplot)

# Alt üçgen matrise korelasyon katsayısı ekleyelim
g.map_lower(sns.kdeplot)

# Grafiki göster
plt.show()
```

**Output Açıklaması**

Bu grafikte:

- Üst üçgen matris, iki farklı değişken arasındaki ilişkiyi gösteren scatter plot'lar içerir. Örneğin, ilk satırın ikinci sütununda total_bill ve tip arasındaki ilişki gösterilir.

- Köşegen (diagonal) matris, her bir değişkenin dağılımını gösteren histogramlar içerir.

- Alt üçgen matris, iki değişken arasındaki yoğunluğu gösteren KDE (Kernel Density Estimation) plot'lar içerir.

PairGrid, bir veri setindeki sayısal değişkenler arasındaki ilişkileri detaylı bir şekilde görmemizi sağlar. Örneğin, bu grafik sayesinde toplam hesap miktarı arttıkça bahşiş miktarının da genellikle arttığını görebiliriz. Ayrıca, her bir değişkenin kendi dağılımını ve yoğunluğunu da gözlemleyebiliriz.

#### facetGrid

Seaborn kütüphanesinde "Grid Plot" olarak adlandırılan görselleştirmeler, birden fazla alt grafik oluşturarak değişkenler arasındaki ilişkileri daha detaylı bir şekilde gözlemlememizi sağlar.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Tips veri setini yükle
tips = sns.load_dataset("tips")

# FacetGrid oluşturma
g = sns.FacetGrid(tips, col="day", row="sex")

# Her bir alt grafiğe bir scatter plot ekleyelim
g.map(sns.scatterplot, "total_bill", "tip")

# Grafiki göster
plt.show()
```

**Output Açıklaması**

Bu grafikte:

- Her bir alt grafik, farklı bir gün ve cinsiyet kombinasyonu için total_bill (toplam hesap) ve tip (bahşiş) arasındaki ilişkiyi gösterir.

- Örneğin, en üst sol köşedeki grafik Perşembe günü için erkeklerin total_bill ve tip arasındaki ilişkisini gösterir.

- Bu tür bir görselleştirme, değişkenler arasındaki ilişkilerin farklı kategoriler veya gruplar için nasıl değiştiğini görmemizi sağlar.

- FacetGrid fonksiyonu oldukça esnektir ve farklı türde grafikler eklememize olanak tanır. Bu sayede, veri setindeki değişkenler arasındaki ilişkileri çok daha detaylı bir şekilde inceleyebiliriz.
