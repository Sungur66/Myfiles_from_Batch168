<p>Veriler datasets/transfusion.data'da saklanr ve RFMTC pazarlama modeline (RFM'nin bir eidi) gre yaplandrlmtr.   

RFM genellikle mteri segmentasyonu iin kullanlr ve mterilerin ne zaman (Recency), ne sklkla (Frequency), ve ne kadar para harcad (Monetary) gibi zellikler zerinden analiz yaplmasn salar. 

Bu zellikler genellikle mteri yaam deeri modellemesi, churn tahmini, mteri segmentasyonu gibi konulara uyarlanr. 

Ancak burada, bu zellikler bir sosyal iyilik konusu olan kan ba iin kullanlmtr.

Bu veri setinde:

**RFMTC Bileenleri**

1. **Recency (R) - Yaknlk   "Recency (months)"**
    - Bu zellik, bir bann son bandan bu yana ne kadar sre getiini temsil eder. Genellikle, son ba daha yakn olan balarn tekrar ba yapma olasl daha yksektir. 
  
2. **Frequency (F) - Sklk  "Frequency (times)"**
    - Bu, bir bann ne sklkla kan ba yaptn gsterir. Yksek sklkta ba yapan kiilerin, gelecekte de ba yapma olasl genellikle daha yksektir.
  
3. **Monetary (M) - Parasal Deer  "Monetary (c.c. blood)"**
    - Bu zellik, bann toplamda ne kadar kan baladn temsil eder. Genellikle, daha yksek miktarda kan balayan balarn deeri daha yksek kabul edilir.

4. **Time (T) - Zaman  "Time (months)"**
    - Bu, bir bann ilk bandan bu yana ne kadar sre getiini gsterir. Bu zellik, bann ba yapma sresi boyunca ne kadar "sadk" olduunu anlamak iin kullanlabilir.

5. **Churn (C) - Ayrlma  "whether he/she donated blood in March 2007"**
    - Bu, bir bann belirli bir zaman diliminde (Mart 2007'de) ba yaplp yapmadn gsterir. Churn, bu rnekte bann o dnemde ba yapmama olasln temsil eder.

**RFMTC'nin Kullanm Alanlar**

1. **Segmentasyon**: Balar, bu zellikler kullanlarak farkl segmentlere ayrlabilir. rnein, yksek "F" ve dk "R" deerine sahip balar "Sadk Balar" olarak etiketlenebilir.

2. **Tahminleme**: Gelecekteki ba olasl, mevcut RFMTC deerleri kullanlarak tahmin edilebilir.

3. **Hedefleme**: zel kampanyalar veya tevikler, belirli ba segmentlerini hedeflemek iin kullanlabilir.

4. **Risk Analizi**: Dk sklkta ve yksek churn oranna sahip balar "Riskli" olarak etiketlenebilir, ve bu balara ynelik zel stratejiler gelitirilebilir.

Bu modelleme teknii, balarn gelecekteki davranlarn anlamak ve onlar daha etkin bir ekilde ynetmek iin olduka kullanldr.

Balarn gelecekte kan ba yapma olasln modellemek iin kullanlabilir.

## Feature Analysis

```python
output_data = []

for col in df.columns:
    
    # If the number of unique values in the column is less than or equal to 5
    if df.loc[:, col].nunique() <= 5:
        # Get the unique values in the column
        unique_values = df.loc[:, col].unique()
        # Append the column name, number of unique values, unique values, and data type to the output data
        output_data.append([col, df.loc[:, col].nunique(), unique_values, df.loc[:, col].dtype])
    else:
        # Otherwise, append only the column name, number of unique values, and data type to the output data
        output_data.append([col, df.loc[:, col].nunique(),"-", df.loc[:, col].dtype])

output_df = pd.DataFrame(output_data, columns=['Column Name', 'Number of Unique Values', ' Unique Values ', 'Data Type'])

output_df
```

## Classify DataFrame Columns into Categorical and Numeric Types

```python
def grab_col_names(dataframe, cat_th=10):

    # cat_cols
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "O"]
    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and
                   dataframe[col].dtypes != "O"]    
    cat_cols = cat_cols + num_but_cat
    

    # num_cols
    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O"]
    num_cols = [col for col in num_cols if col not in cat_cols]

    print(f"Features: {dataframe.shape[1]}")
    print(f'Number of Categorical Features: {len(cat_cols)}')
    print(f'Number of Numeric Features: {len(num_cols)}')
    print(f"Categorical Features: {cat_cols}") 
    print(f"Numeric Features: {num_cols}")
    
    return cat_cols, num_cols
```

```python
cat_cols, num_cols = grab_col_names(df)
```

## DataFrame Summary Statistics

```python
def summary(df, pred=None):
    obs = df.shape[0]
    Types = df.dtypes
    Counts = df.apply(lambda x: x.count())
    Min = df.min()
    Max = df.max()
    Uniques = df.apply(lambda x: x.unique().shape[0])
    Nulls = df.apply(lambda x: x.isnull().sum())
    print('Data shape:', df.shape)

    if pred is None:
        cols = ['Types', 'Counts', 'Uniques', 'Nulls', 'Min', 'Max']
        str = pd.concat([Types, Counts, Uniques, Nulls, Min, Max], axis = 1, sort=True)

    str.columns = cols
    print('___________________________\nData Types:')
    print(str.Types.value_counts())
    print('___________________________')
    return str

summary(df)
```

## Check and Remove Duplicate Rows

```python
def duplicate_values(df):
    print("Duplicate check...")
    num_duplicates = df.duplicated(subset=None, keep='first').sum()
    if num_duplicates > 0:
        print("There are", num_duplicates, "duplicated observations in the dataset.")
        df.drop_duplicates(keep='first', inplace=True)
        print(num_duplicates, "duplicates were dropped!")
        print('*' * 100)
    else:
        print("There are no duplicated observations in the dataset.")
```

```python
duplicate_values(df)
```

```python
duplicate_values(df)
```

## Missing Value Analysis in DataFrame

```python
def missing_values(df):
    missing_number = df.isnull().sum().sort_values(ascending = False)
    missing_percent = (df.isnull().sum() / df.isnull().count()).sort_values(ascending = False)
    missing_values = pd.concat([missing_number, missing_percent], axis = 1, keys = ['Missing_Number', 'Missing_Percent'])
    return missing_values[missing_values['Missing_Number'] > 0]
```

```python
missing_values(df)
```

## Column Value Distribution Analysis

```python
def value_cnt(df, column_name):
    vc = df[column_name].value_counts()
    vc_norm = df[column_name].value_counts(normalize=True).round(3)

    vc = vc.rename_axis('workclass').reset_index(name='counts')
    vc_norm = vc_norm.rename_axis('workclass').reset_index(name='norm_counts')

    df_result = pd.concat([vc['workclass'], vc['counts'], vc_norm['norm_counts']], axis=1)
    
    return df_result
```
