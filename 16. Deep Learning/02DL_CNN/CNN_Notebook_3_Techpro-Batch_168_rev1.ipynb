## Preparing the Data for the model

```python
uninfected_cell.max()
```

```python
para_img.max()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
batch_size = 32
```

```python
image_gen = ImageDataGenerator(rotation_range=20, # rotate the image 20 degrees
                               width_shift_range=0.10, # Shift the pic width by a max of 10%
                               height_shift_range=0.10, # Shift the pic height by a max of 10%
                               #rescale=1/255, # Rescale the image by normalzing it.
                               shear_range=0.1, # Shear means cutting away part of the image (max 10%)
                               zoom_range=0.1, # Zoom in by 10% max
                               horizontal_flip=True, # Allo horizontal flipping
                               fill_mode='nearest' # Fill in missing pixels with the nearest filled value
                              )
```

Görüntüyü döndürdüğünüzde veya değiştirdiğinizde, bazı yerlerde boşluklar (siyah veya boş pikseller) oluşabilir. fill_mode='nearest', bu boşlukları, boşluğun hemen yanındaki pikselin rengiyle doldurur. Yani, boşluğun yanındaki pikselin rengini alır ve boşluğu bu renkle doldurur. Bu, görüntünün daha doğal görünmesini sağlar.

# Creating the Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dropout, Flatten, Dense, Conv2D, MaxPooling2D
```

```python
batch_size = 32
```

```python
model = Sequential()

model.add(Conv2D(filters=32, kernel_size=(3,3),input_shape=image_shape, padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=32, kernel_size=(3,3), padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=32, kernel_size=(3,3), padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())


model.add(Dense(32))
model.add(Activation('relu'))


model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.summary()
```

## Early Stopping

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss',patience=10)
```

## Training the Model

```python
#help(image_gen.flow_from_directory)
```

```python
train_image_gen = image_gen.flow_from_directory(train_path,
                                               target_size=(130,130),
                                               color_mode='rgb',
                                               batch_size=batch_size,
                                               class_mode='binary', shuffle=True)
```

```python
test_image_gen = image_gen.flow_from_directory(test_path,
                                               target_size=(130,130),
                                               color_mode='rgb',
                                               batch_size=batch_size,
                                               class_mode='binary',shuffle=True)
```

# MODEL TRAINING

```python
results = model.fit(train_image_gen,epochs=20,
                    validation_data=test_image_gen,
                    callbacks=[early_stop])
```

# Evaluating the Model

predictions = pred_probabilities > 0.5 ifadesi, belirli bir tahmin olasılıkları dizisi (pred_probabilities) içindeki her değerin, belirli bir eşik değeri aşıp aşmadığını kontrol eder. Bu durumda, eşik değeri 0.5 olarak belirlenmiştir.

Kodun ne yaptığını daha ayrıntılı bir şekilde açıklayacak olursak:

pred_probabilities: Bu, genellikle bir modelin her bir örnek için ürettiği tahmin olasılıklarını içeren bir dizi veya liste olabilir. Bu olasılıklar, genellikle 0 ile 1 arasında bir değer alır.

> 0.5: Bu operatör, pred_probabilities dizisindeki her bir olasılık değerini 0.5 ile karşılaştırır. Eğer bir değer 0.5'ten büyükse, bu karşılaştırma True (Doğru) olarak değerlendirilir; eğer 0.5'ten küçük veya eşitse, False (Yanlış) olarak değerlendirilir.

predictions: Bu, yukarıdaki karşılaştırmanın sonucu olarak oluşturulan yeni bir dizi veya liste olur. Bu dizi, her bir örneğin belirlenen eşik değerini aşıp aşmadığına dair boolean (Doğru/Yanlış) değerler içerir.

Bu tür bir kod parçası genellikle ikili sınıflandırma problemlerinde kullanılır, burada model bir örneğin belirli bir sınıfa ait olup olmadığını tahmin eder ve bu ifade bu tahminleri belirli bir eşik değere göre ikili (evet/hayır, pozitif/negatif vb.) sonuçlara dönüştürmek için kullanılır.
