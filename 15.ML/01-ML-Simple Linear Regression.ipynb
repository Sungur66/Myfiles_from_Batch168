# Spliting the dataset into X (independent variables) and y (dependent variables)

y = b1 * X + b0
### y = bağımlı değişken = target variable = dependent variable
### b1 = katsayı/eğim, slope -∞ < 0 < +∞
### X = bağımsız değişken /feature, independent variable
### b0 = otonom parametre / intercept /başlangıç parametresi -∞ < 0 < +∞

```python
X = df["total_spend"] # feature
y = df["sales"]  # target
```

```python
# icine ind. ve dep. variable verdigimde ve derecesini belirttimde danklemin katsayilarini vercek
# sadece simple linear reg icin

np.polyfit(X, y, deg=1)

""" array([0.04868788, 4.24302822])
y = b1 * X + b0
sonucun ilk degeri X in katsayisi yani (b1) - slope 
sonucun ikinci degeri intercept olur - b0
"""
```

```python
slope, intercept = np.polyfit(X, y, deg=1)
print("slope    :", slope)
print("intercept:", intercept)
```

```python
b1, b0 = np.polyfit(X, y, deg = 1)
print("b1 :", b1)
print("b0 :", b0)
```

```python
# y_pred i formulize edelim.. y_pred = b1*x + b0

y_pred = b1*X + b0
y_pred
```

##### Residual term(e) = observed value - predicted value

y = 0.049 * X + 4.24 + ε (hata parametreleri toplamı gerçek değeri bulmak için- çizgiden gerçek değer arasındaki mesafenin değeri) residual errors

actual                 = y               = gerçek değer

predicted              = y_pred          = tahmin değeri

residual               = y - y_pred      = gerçek değer ve tahmin edilen arasındaki fark (erorler)

LSE (Least Square Eror) = (y - y_pred)^2  = gerçek değer ve tahmin edilen arasındaki farkın karesi=En küçük kareler

```python
values = {"actual": y, "predicted":y_pred, "residual": y - y_pred, "LSE": (y - y_pred)**2}
df_2 = pd.DataFrame(values)
df_2
```

```python
df_2.residual.sum().round()   # residual larin toplam degerini yuvarlar
```

```python
df_2.LSE.sum()
```

# How can we improve our predictions?

more feature, hyperparameter, polinomial, bias-variance

overfitting x den x**2, x**3, x**4 yapmak complexity yi artirir  ancak variance artar yani tahmin zorlasir

y = b1*x + bß  -> simple linear

y = a1 * x**3 + a2 * x**2 + a3 * x + a0  -> 3 degree polinominal
polynominal olarak tahmini iyileştirebiliriz.

```python
a = np.polyfit(X,y, deg=3)
a
"""
# array([ 3.07615033e-07, -1.89392449e-04,  8.20886302e-02,  2.70495053e+00])
# fonksiyonun döndürdugu sonuclar polinom katsayilarini verir. eger cikti [a,b,c,d] 
seklindeyse, bu, bulunan en iyi uyan polinomun y = ax^3 + bx^2 + cx + d 
"""
```

```python
a1 = np.polyfit(X, y, deg=3)[0]
a2 = np.polyfit(X, y, deg=3)[1]
a3 = np.polyfit(X, y, deg=3)[2]
a0 = np.polyfit(X, y, deg=3)[3]
```

```python
#polynom fonksiyon için tahminleri almış oldum
predicted_sales_poly = a1 * potential_spend**3 + a2 * potential_spend**2 + a3 * potential_spend + a0
predicted_sales_poly
```

```python
plt.plot(potential_spend, predicted_sales_poly, color="r");
```

```python
plt.plot(potential_spend, predicted_sales_poly, color="r")
sns.scatterplot(x="total_spend", y="sales", data=df);
```
