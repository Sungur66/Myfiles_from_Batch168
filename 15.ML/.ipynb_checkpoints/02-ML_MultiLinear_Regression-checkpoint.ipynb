skewness icin grafik 

```python
# skewness tespiti icin bir döngü

for i in df.drop(columns = "sales").columns:
    print(i, skew(df[i]))
    
# sales sütunu haric tüm feature larin skewness degerini hesaplar
# carpiklik 0 ise normal dagilim  (-0.5 ile +0.5 arasi dagilimlar)
# carpiklik > 0 ise sagdan carpik - kuyruk sagda
# carpiklik < 0 ise soldan carpik- kuyruk solda
```

```python
for j in df.drop(columns = "sales").columns:     #"sales" sütunu dataframe'den kaldırılıyor
    g = sns.distplot(df[j], color= "b", label= "Skewness : %.2f"%(df[j].skew()))   #  j sütununa ait veri dağılımı grafiği çiziliyor. grafiğin üzerinde gösterilen etiketi oluşturur. Etiket, o sütunun çarpıklık değerini (skewness) içerir
    g = g.legend(loc= "best")   #  grafiğin üzerine eklenen etiketin konumunu en iyi yere yerleştirmes
    plt.show()
```

# Train-Test Split

```python
# !pip install scikit-learn
# !pip install sklearn

#  X : Feature    y : Target variable artik bunlar buseklide ifade edilecek
# X : Feature (TV-radio-newspaper)    y : Target (sales)
```

```python
# pip install scikit-learn       bu calisti esas
```

```python
X = df.drop(columns = "sales")  # X = df[["TV", "radio", "newspaper"]] alternativ kullanim
y = df["sales"]     # X ler feature olarak, y target olarak label landi

print(X, y)
```

```python
from sklearn.model_selection import train_test_split

# veri setimizi randomly train ve test setlerine ayirir
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
df.sample(15)
```

```python
print("X Train features shape: {}\ny Train features shape: {}\nX Test features shape : {}\nY Test features shape : {}".format
      (X_train.shape, y_train.shape, X_test.shape, y_test.shape))

# train ve test setlerinin shape lerine bakiyoruz
```

```python
#Print("Train features shape: ", X_train.shape)
#Print("Train target shape: ", y_train.shape)
#Print("Test features shape: ", X_test.shape)
#Print("Test target shape: ", y_test.shape)
```

# Model Fitting and Compare Actual and Predicted Labels

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)  
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
model.coef_   # y = ax1 + bx2 + cx3 + b0      esitligindeki a,b, c
```

```python
model.feature_names_in_ # yukardaki katsayilarin karsilik geldigi feature lar
```

```python
model.intercept_  # bu da b0
```

```python
# herhangi bir satirdaki degerler icin tahmin yapalim / manuel olarak y_pred hesaplandi
sum(X_test.loc[95]*model.coef_) + model.intercept_
```

```python
y_test.loc[95]
```

```python
my_dict = {"Actual": y_test, "pred": y_pred, "residual": y_test - y_pred}
```

```python
comparing = pd.DataFrame(my_dict)
comparing
```

```python
# 25 adet random sonucu alip residuallari görsel olarak degerlendirelim 
result_sample = comparing.head(25)
result_sample
```

```python
result_sample.plot(kind="bar", figsize = (15,9))
plt.show()
```

# Error Metrics

```python
# modelin basarisini ölcme
# performans degerlendirmek icin error metric leri yüklemilyiz

from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
```

```python
R2_score = r2_score(y_test, y_pred)
R2_score
```

```python
# MAE hesabi, mean absolute error / bu degerin 0 a yakin olmasi makbul

mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
# MSE hesabi, mean squared error / bu degerin 0 a yakin olmasi makbul

mse = mean_squared_error(y_test, y_pred)
mse
```

```python
#  root mean squared error (mse nin kare kökü)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
rmse
```

```python
# Ortalama Hata orani

sales_mean = df["sales"].mean()
sales_mean
```

```python
# MAE Hata Orani - degerin 0 a yakin olmasi iyi bir tahmin modeli oldugunu gösterir. 
# MAE için = Ortalama mutlak hata, satışların ortalama değerinin yaklaşık %10.78'i kadardır.
mae / sales_mean   
```

```python
# RMSE Hata Orani
rmse / sales_mean
```
