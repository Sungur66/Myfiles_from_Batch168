# AdaBoost (Adaptive Boosting)

Ensemble yntemleri, bir grup temel reniciyi bir araya getirerek daha gl bir model oluturan tekniklerdir. Outlier'lar (aykr deerler), veri setinde genellikle dier gzlemlerden nemli lde farkl olan nadir veri noktalarn ifade eder. Ensemble yntemlerinin outlier'lar zerindeki etkisi, kullanlan renici trne ve ensemble ynteminin zelliklerine bal olarak deiebilir. te baz genel gzlemler:

Rastgele Ormanlar:

Rastgele ormanlar outlier'lar genellikle daha iyi tolere eder. nk her bir aa farkl bir alt kme zerinde eitildii iin, bir aacn yanl bir ekilde outlier' etkilemesi dier aalar zerinde genellikle zayf bir etkiye sahip olur.

Boosting Yntemleri (AdaBoost, Gradient Boosting vb.):

Boosting yntemleri, nceki renicilerin hatalarn dzeltmeye odakland iin, outlier'lar bu yntemlerde daha fazla etkili olabilir. zellikle bir nceki renicinin yanl tahmin ettii rnekler, bir sonraki renici tarafndan daha fazla vurgulanabilir.

Bagging Yntemleri (Bootstrap Aggregating):

Bagging yntemleri, renicileri bamsz olarak eittii ve ardndan tahminleri birletirdii iin outlier'lar genellikle daha iyi tolere eder. Ancak, her reniciyi benzer ekilde eitmek, outlier'larn etkisini azaltabilir.

Ensemble Ynteminin Genel Etkisi:

Ensemble yntemleri, genelde outlier'larn etkisini azaltabilir, nk birok reniciyi bir araya getirerek genel bir tahmin yaparlar. Ancak, outlier'lar modelin genel performansn olumsuz ynde etkileyebilir, zellikle veri setinde belirgin bir ekilde yanltc ya da hatalysa.
Outlier'larn etkisi genellikle veri seti zelliklerine ve kullanlan ensemble ynteminin zelliklerine bal olarak deiir. Bu nedenle, renme srecinde ve modelin performansn deerlendirirken outlier'lar dikkate almak nemlidir.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (10,6)
import warnings
warnings.filterwarnings("ignore")
warnings.warn("this will not show")
```

```python
df = pd.read_csv("advertising2.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("Clicked on Ad", axis=1)
y = df["Clicked on Ad"]
```

```python
cat = X.select_dtypes("object").columns
cat
```

```python
from sklearn.preprocessing import OrdinalEncoder

enc = OrdinalEncoder()
X[cat] = enc.fit_transform(X[cat])
X.head(20)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, recall_score, precision_score, f1_score, roc_auc_score
from sklearn.model_selection import cross_val_score, cross_validate
```

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
ada_model= AdaBoostClassifier(n_estimators= 50, random_state=42) #DecisionTreeClassifier(max_depth=1)
```

```python
ada_model.fit(X_train, y_train)
```

```python
eval_metric(ada_model, X_train, y_train, X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score, cross_validate
model = AdaBoostClassifier(n_estimators=50, random_state=42)

scores = cross_validate(model, X_train, y_train, scoring = ["accuracy", "precision", "recall", "f1"], cv = 10)
df_scores = pd.DataFrame(scores, index = range(1, 11))
df_scores.mean()[2:]
```
