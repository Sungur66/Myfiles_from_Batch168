# Principal Component Analysis-Unsupervised

Principal Component Analysis (PCA) makine öğrenmesinde ve istatistikte yaygın olarak kullanılan bir boyut azaltma tekniğidir. Veri setindeki **değişkenler arasındaki korelasyonu temel alarak**, orijinal veri setini daha az sayıda bileşene indirger. Bu bileşenler, veri setindeki maksimum varyansı korumaya çalışırken orijinal veri setinin "ana özelliklerini" yakalamaya çalışır.

- Bu yöntem, çok boyutlu veri setlerindeki değişkenlikleri en iyi şekilde özetleyen bileşenleri belirlemek için kullanılır.  

- 20 kg nar yemek mi yoksa bunun 2 litre nar suyunu içmek mi ?

- çok feature ile çalışmanın dezavantajları...mesela görselleştirmedeki ve görseli okumadaki dezavantajlar

- Multicollinearity sorunu çözlüyor

- Lineer ilişki olan feature'lar olduğunda başarılı olur

**PCA Uygulama Adımları:**

Veriyi Standartlaştırın: Özellikler farklı ölçeklere sahip olabileceği için (örn., motor gücü "hp" cinsinden, ağırlık "kg" cinsinden olabilir), veriyi standartlaştırmalısınız.

Kovaryans Matrisini Hesaplayın: Özellikler arasındaki ilişkiyi belirlemek için.

Özdeğerler ve Özvektörleri Hesaplayın: Bu, kovaryans matrisinin özdeğerleri ve özvektörleri olarak elde edilir.

Ana Bileşenleri Seçin: Özdeğerlere göre ana bileşenleri sıralayarak ilk 2 veya 3 ana bileşeni seçebilirsiniz.

Veriyi Yeni Bileşenlere Dönüştürün: Orijinal veriyi, seçtiğiniz ana bileşenler üzerinden yeni bir veri kümesine dönüştürün.

## PCA

```python
from sklearn.decomposition import PCA
```

Burada n_compnents degeri gireceğiz.

PCA'da n_components sayısını seçerken dikkate alınması gereken birkaç önemli faktör vardır:

-Varyansı Açıklama Oranı: n_components sayısını belirlerken, veri setinizdeki toplam varyansın ne kadarının korunmasını istediğinize karar vermelisiniz. Her bileşen, veri setindeki toplam varyansın bir kısmını açıklar. Genellikle, veri setindeki toplam varyansın büyük bir yüzdesini (örneğin, %80) açıklayacak şekilde n_components sayısını seçersiniz.

-Model Performansı: PCA'nın uygulandığı makine öğrenimi modelinin performansını optimize etmek için n_components seçimi yapılabilir. Farklı bileşen sayıları deneyerek modelin doğruluğunu veya başka performans metriklerini karşılaştırabilirsiniz.

-Görselleştirme: Eğer PCA'yı verilerinizi görselleştirmek için kullanıyorsanız, genellikle iki veya üç bileşen yeterli olacaktır, çünkü bu, verileri iki veya üç boyutta kolayca görselleştirebilmenizi sağlar.

-Hesaplama Maliyeti: Daha az bileşen kullanmak, genellikle daha düşük hesaplama maliyeti anlamına gelir. Büyük veri setleriyle çalışırken, bu, önemli bir faktör olabilir.

```python
pca = PCA(n_components=2)   # 2 bileşen seçildi
```

```python
principal_components = pca.fit_transform(scaled_X)   # 30 feature'ı 2 bileşene dönüştürdü
```

```python
principal_components

# 2 sütun halinde 2 bileşene idirdik 30 feature ı
```

PCA'da bir bileşenin içindeki özelliklerin ağırlıklarına "yük" (loading) denir. Bu yükler, bileşenin o özellik tarafından ne kadar temsil edildiğini gösterir.

Bir özellik için bir bileşende yük değeri pozitifse ve diğer bir bileşende negatifse, bu özelliklerin bu iki bileşende farklı yönlerde olduğunu gösterir.

Örneğin, bir özellik (feature) için:

İlk bileşende 0.21 değeri: Bu özellik, bu bileşeni oluşturan varyansın pozitif yönünde etkili. Yani, bu bileşenin değeri arttıkça, bu özellik de artar.

İkinci bileşende -0.19 değeri: Bu özellik, bu bileşeni oluşturan varyansın negatif yönünde etkili. Bu da demektir ki, bu bileşenin değeri arttıkça, bu özellik azalır.

Veri Yorumlama: PCA sonrası bileşenler, orijinal özelliklerin doğrusal kombinasyonları olduğu için, bu bileşenlerin doğrudan yorumlanması genellikle zordur. Ancak, özelliklerin bileşenlerdeki yükleri (ağırlıkları), hangi özelliklerin belirli bir bileşen için daha önemli olduğunu gösterir.

## K-Means Algorithm Application

burada KMeans üzerinden bir değerlendirme yapalım

Öncelikle 2 bileşenin inertia değerine bkalım. Inertia için minimum değer iy olarak değerlendirilir. Kümenin centroid etrafındaki yoğunlaşmasını gösterir

### Hopkins test

```python
from sklearn.neighbors import BallTree
import numpy as np
import pandas as pd
def hopkins(data_frame, sampling_size):
    """Assess the clusterability of a dataset. A score between 0 and 1, a score around 0.5 express
    no clusterability and a score tending to 0 express a high cluster tendency.
    Parameters
    ----------
    data_frame : numpy array
        The input dataset
    sampling_size : int
        The sampling size which is used to evaluate the number of DataFrame.
    Returns
    ---------------------
    score : float
        The hopkins score of the dataset (between 0 and 1)
    Examples
    --------
    >>> from sklearn import datasets
    >>> from pyclustertend import hopkins
    >>> X = datasets.load_iris().data
    >>> hopkins(X,150)
    0.16
    """
    if type(data_frame) == np.ndarray:
        data_frame = pd.DataFrame(data_frame)
    # Sample n observations from D : P
    if sampling_size > data_frame.shape[0]:
        raise Exception(
            'The number of sample of sample is bigger than the shape of D')
    data_frame_sample = data_frame.sample(n=sampling_size)
    # Get the distance to their neirest neighbors in D : X
    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(data_frame_sample, k=2)
    data_frame_sample_distances_to_nearest_neighbours = dist[:, 1]
    # Randomly simulate n points with the same variation as in D : Q.
    max_data_frame = data_frame.max()
    min_data_frame = data_frame.min()
    uniformly_selected_values_0 = np.random.uniform(min_data_frame[0], max_data_frame[0], sampling_size)
    uniformly_selected_values_1 = np.random.uniform(min_data_frame[1], max_data_frame[1], sampling_size)
    uniformly_selected_observations = np.column_stack((uniformly_selected_values_0, uniformly_selected_values_1))
    if len(max_data_frame) >= 2:
        for i in range(2, len(max_data_frame)):
            uniformly_selected_values_i = np.random.uniform(min_data_frame[i], max_data_frame[i], sampling_size)
            to_stack = (uniformly_selected_observations, uniformly_selected_values_i)
            uniformly_selected_observations = np.column_stack(to_stack)
    uniformly_selected_observations_df = pd.DataFrame(uniformly_selected_observations)
    # Get the distance to their neirest neighbors in D : Y
    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(uniformly_selected_observations_df, k=1)
    uniformly_df_distances_to_nearest_neighbours = dist
    # return the hopkins score
    x = sum(data_frame_sample_distances_to_nearest_neighbours)
    y = sum(uniformly_df_distances_to_nearest_neighbours)
    if x + y == 0:
        raise Exception('The denominator of the hopkins statistics is null')
    return x / (x + y)[0]
```

```python
hopkins(component_df, component_df.shape[0]) # test sonucu 0 yakınsa clusters için uygundur 
```

```python
# inertia hesabı

from sklearn.cluster import KMeans

ssd = []

K = range(2,10)

for k in K:
    model = KMeans(n_clusters =k, random_state=42)
    model.fit(component_df)
    ssd.append(model.inertia_)
```

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error") 
plt.title("elbow method");

# alttaki, grafikte sert düşüşü arıyoruz
# daha önce yaptıgımız ara işlemleri altta yapacagız tekrar
```

yellowbrikck 4 dedi, silhouette bakalım

```python
from sklearn.metrics import silhouette_score

range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(component_df)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(component_df, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
```

Silhouette için 2 hariç diğerleri birbirine yakın. 2-3-4 seçileiblir. bu datada n_cluster=2 seçilerek kanser yada değil gibi bir clusterlama yapılabilir, yada kanser evrelerine göre bir sıralama yapılabilir 3 ve 4 component seçilirse. belkide cluster ı canser evrelerine göre classlamak uygun olabilir. burada kanser veya değil diye bir sınıflandorma  isteniyorsa 2 bileşenle devam edilir
