# ACF ve PACF Grafikler

```python
# VEriler duraan deilse bu garafiklerin kontrol ile serinin duraanl hakknda karmlar yapabiliriz
# Zaman serileri analizinde verilerin otokorelasyon (ACF) ve ksm otokorelasyon grafiinin (PACF) incelenmesi nemli bir admdr. \\ 
# \\ Model terimlerinin saysn belirlemek iin Otokorelasyon fonksiyonu (ACF) ve ksmi otokorelasyon fonksiyonu (PACF) \\
# \\grafiklerinden yararlanacaz
```

```python
# otokorelasyon ve ksm otokorelasyon grafikleri (ACF-PACF): 

fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(20,5), dpi= 70)

plot_acf(data_daily.Tuketim.tolist(), ax=ax1, lags=48)
plot_pacf(data_daily.Tuketim.tolist(), ax=ax2, lags=48, method="ywm")

plt.show()
```

```python
# Verilerin izgi grafii, zaman serileri bileenleri ve ACF-PACF grafiinden yola karak \\ 
# \\ serinin duraan olmadn syleyebiliriz.
```

Grsel olarak yaplan Duraanlk durumu iin (Trend ve Mevsimsellikden dolay) bu tespitlerini dogrulamak zere Dickey-Fuller ve KPSS hipotez testleri kullanlr


1- Augmented Dickey-Fuller Test 
* Sfr Hipotez H0 : Seri Duraan Deildir
* Alternati Hipotez H1 : Seri Duraandr (Birim Kk Yoktur)

2- Kwiatkowski-Phillips-Schmidt-Shin Test
* Sfr Hipotez H0 : Seri Duraandr (Birim Kk Yoktur)
* Alternati Hipotez H1: Seri duraan deildir

```python
# ADF Testi:

def adf(data):
    adf_test = sm.tsa.adfuller(data, autolag='AIC')
    print(f'ADF Statistic: {adf_test[0]}')
    print(f'p-value: {adf_test[1]}')
    print("Seri Duraan Deildir") if adf_test[1] > 0.05 else print("Seri Duraandr(Birim Kk Yoktur)")

# KPSS Testi:

def kpss(data):
    kpss_test = sm.tsa.kpss(data)
    print('KPSS Statistic: %f' % kpss_test[0])
    print('p-value: %f' % kpss_test[1])
    print("Seri Duraandr") if kpss_test[1] > 0.05 else print("Seri duraan deildir(Seri Birim Kk erir)")

adf(data_daily)
print()
kpss(data_daily)
```

```python
# p-deerinin 0.05'ten (ve hatta 0.01'den) kk olduunu gz nnde bulundurarak, 
# bu zaman serisinin %95 (ve %99) gven seviyesinde duraan olduunu syleyebiliriz.
# ADF istatistii negatif ise hipotez duraanlk lehine reddedilir.
```

```python
# ADF testi bir birim kk bulamazsa, ancak KPSS testi bulursa, seri fark duraandr ve yine de fark alnmas gerekir.
```

```python
# Testlerden biri bile birim kkn varlna iaret ediyorsa, seri duraandr diyemeyiz.
```

Alttaki blmde: 
    Serinin duraan olmadna emin olduktan sonra fark yntemi ile serinin birinci dereceden farkn alarak serinin duraan hale gelip gelmediini test edelim.
    Ayrca baka metotlarda vardr. ders notunda bunlar belirtildi.

```python
data_daily_diff = data_daily.copy()
data_daily_diff['Tuketim'] = data_daily['Tuketim'] - data_daily['Tuketim'].shift(1)
data_daily_diff = data_daily_diff.dropna()
```

```python
#  birinci dereceden fark alnan serinin ACF ve PACF grafikleri:

fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(20,5), dpi= 70)

plot_acf(data_daily_diff.Tuketim.tolist(), ax=ax1, lags=48)
plot_pacf(data_daily_diff.Tuketim.tolist(), ax=ax2, lags=48, method="ywm")

plt.show()
```

```python
#  birinci dereceden fark alnan serinin ADF-KPSS testleri: 

adf(data_daily_diff)
print()
kpss(data_daily_diff)
```

```python
#Serimiz artk duraan hale gelmi olsa bile hala ACF grafiinde dalgalanmalar yani mevsimselliin olduunu ve gecikme saylarnn tekrarlarn incelediimiz zaman \\
# \\ periyodun 7 olduunu tespit edebiliyoruz. Ayrca PACFdeki 7. gecikmenin de snrlar dnda kaldn gryoruz.
```

```python
# Periyot 7 olarak belirledikten sonra birinci dereceden mevsimsel fark alyoruz:
# birinci dereceden fark alnan seri zerinde mevsimsel fark uygulanmas: 

data_daily_diff2 = data_daily_diff.diff(7).dropna()
```

```python
#  ACF ve PACF plot

fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(20,5), dpi= 70)

plot_acf(data_daily_diff2.Tuketim.tolist(), ax=ax1, lags=48)
plot_pacf(data_daily_diff2.Tuketim.tolist(), ax=ax2, lags=48)

plt.show()
```

Yukardaki ilemler sonucunda bylece serimizi trend ve mevsimsel etkiden arndrmay baardk.

# Modelleme Aamas

 Modelleme aamasna gemeden nce, yaptmz analizler dorultusunda model parametrelerini manuel olarak belirleyemeye alalm.
 

Hem trend hem mevsimsel fark alnarak duraan hale geldii iin d = 1 ve D = 1 eklinde tanmlayacaz.

ACF grafiinde 1,2 ve 7' de nemli bir gecikme vardr. Bu nedenle MA(q) parametresi 2 ve mevsimsel MA(Q) parametresi 1 olabilir.

PACF grafiinde ise 1,2 ve 7,14,21de nemli bir gecikme vardr. AR(p) parametresi 2 olabilir. Mevsimsel AR(P) parametresinin ise aslnda bu durumda 0 alnmas tavsiye ediliyor.

Baz kaynaklarda, mevsimsel gecikmedeki otokorelasyon negatifse, modele mevsimsel MA(Q) terimi eklenip mevsimsel AR(P) teriminin atlanmas yani 0 alnmas neriliyor. Fakat   denemelerde AR(P)=1 almak daha dk hata oran ve yksek R Kare deeri verdii iin biz burada AR(P) terimini 1 olarak alacaz.

Buradan yola karak;

ARIMA(2,1,2) ve SARIMA(2,1,2)x(1,1,1,7) olacak ekilde notasyonumuzu oluturuyoruz.

```python

# veri setinin eitim ve test seti olarak ayrtrlmas : 
# 1 Ocak 2017 - 31 Temmuz 2022 eitim seti, 1 Austos 2022 - 31 Austos 2022 ise test seti olacak ekilde ayryoruz. 

train_data = data_daily[:'2022-07-31']
test_data = data_daily['2022-08-01':'2022-08-31']
pred_df = test_data.copy()
```

Her model iin ayn aralkta bir tahminleme ilemi yapacaz sonrasnda sonular test setindeki gerek deerler ile karlatracaz.

## ARIMA modellemesi

```python
#ARIMA(2,1,2) modelini statsmodels paketini kullanarak oluturalm ve belirlediimiz aralk iin bir tahminleme \\
# \\ gerekletirelim. Ardndan gerek deerler ile tahmin edilen deerleri bir grsel zerinde gsterelim.
```

```python
# ARIMA (2,1,2)
#Verileri eitim ve test seti olarak ikiye ayryoruz

arima_model = sm.tsa.arima.ARIMA(np.asarray(train_data),order=(2,1,2))
arima_model_fit = arima_model.fit()
arima_pred = arima_model_fit.forecast(len(test_data),disp=False)

pred_df["arima_pred"] = arima_pred.round(2)
plt.figure(figsize=(22,4))
sns.lineplot(data = pred_df[['Tuketim', 'arima_pred']])
```

ARIMA modeli grafikten anlayacamz zere pek iyi almad nk verilerimiz mevsimsellik ieriyor ve ARIMA tarafndan iyi tahmin edilemiyor. Bu tr durumlarda daha nceki yazmda bahsettiim gibi SARIMA modeli daha iyi alacaktr

 ## SARIMA Modellemesi

ayn tahmin ilemini SARIMA zerinde gerekletirelim. 
SARIMA (2,1,2) (1,1,1,7)

```python
# SARIMA (2,1,2) (1,1,1,7)

sarima_model = SARIMAX(np.asarray(train_data), order=(2, 1, 2), seasonal_order=(1,1,1,7))
sarima_model_fit = sarima_model.fit()
sarima_pred = sarima_model_fit.forecast(len(test_data))

pred_df["sarima_pred"] = sarima_pred.round(2)
plt.figure(figsize=(22,4))
sns.lineplot(data = pred_df[['Tuketim', 'sarima_pred']])
```

```python
# SARIMA modelinin ARIMA modelinden daha iyi sonu verdiini grafikten gzlemleyebiliyoruz.
```

## PROPHET Modellemesi

```python
# Prophet algoritmas girdi olarak ds ve y deikenleriyle alt iin eitim ve test setimizi buna gre ayarlyoruz
```

ds:  Bu stun, tarihleri ierir. Prophet iin bu stunda olan tarihler genellikle gnlk frekansta olmaldr, ancak bu zorunlu deildir. Tarihler YYYY-MM-DD biiminde (rnein "2021-01-01") ya da zaman damgas (timestamp) biiminde olabilir.

y: Bu stun, tahmin etmek istediiniz gzlemlenen deerleri ierir. rnein, bir web sitesinin gnlk ziyareti saysn tahmin etmek istiyorsanz, y stunu bu ziyareti saylarn ierecektir.

```python
# PROPHET

prophet_df = data_daily.reset_index()
prophet_df.columns = ["ds","y"]

train_data_prophet = prophet_df[prophet_df['ds'] < '2022-08-01']
test_data_prophet = prophet_df[(prophet_df['ds'] >= '2022-08-01') & (prophet_df['ds'] <= '2022-08-31')]

prophet_model = Prophet()
prophet_model.fit(train_data_prophet)
model_predict = prophet_model.predict(test_data_prophet)

pred_df["prophet_yhat"] = model_predict["yhat"].values.round(2)
pred_df["prophet_yhat_lower"] = model_predict["yhat_lower"].values.round(2)
pred_df["prophet_yhat_upper"] = model_predict["yhat_upper"].values.round(2)

plt.figure(figsize=(22,4))
sns.lineplot(data = pred_df[['Tuketim', 'prophet_yhat']])
```

```python
# altta ise yhat, yhat_lower ve yhat_upper deerlerini ieren grafik gorunuyor
```

yhat: Bu, modelin tahmin ettii ana deerdir. Dier bir deyile, bu deer modelinizin belirli bir ds (tarih) deeri iin zaman serisinin ne olacan tahmin ettii deerdir.

yhat_lower ve yhat_upper: Bu deerler, tahminlerin belirsizlik araln ifade eder. Prophet, tahminlerini olutururken belirsizlii de dikkate alarak, gerek deerin bu aralk ierisinde olma olasln yksek bir gvenle tahmin eder. rnein, eer yhat_lower 50 ve yhat_upper 60 ise, bu modelin gerek deerin bu aralkta olacan yksek bir gvenle tahmin ettii anlamna gelir.

```python
plt.figure(figsize=(22,4))
sns.lineplot(data = pred_df[['Tuketim', 'prophet_yhat', 'prophet_yhat_lower','prophet_yhat_upper']])
```

# Model Evaluation

```python
# Modeli deerlendirmek iin MAE, RMSE ve R2 metriklerini kullanacaz
```

```python
prophet_mape = mean_absolute_percentage_error(pred_df["Tuketim"].values, pred_df["prophet_yhat"].values)
prophet_r = r2_score(pred_df["Tuketim"].values, pred_df["prophet_yhat"].values)
prophet_rmse = math.sqrt(mean_squared_error(pred_df["Tuketim"].values, pred_df["prophet_yhat"].values))

arima_mape = mean_absolute_percentage_error(pred_df["Tuketim"].values, pred_df["arima_pred"].values)
arima_r = r2_score(pred_df["Tuketim"].values, pred_df["arima_pred"].values)
arima_rmse = math.sqrt(mean_squared_error(pred_df["Tuketim"].values, pred_df["arima_pred"].values))

sarima_mape = mean_absolute_percentage_error(pred_df["Tuketim"].values, pred_df["sarima_pred"].values)
sarima_r = r2_score(pred_df["Tuketim"].values, pred_df["sarima_pred"].values)
sarima_rmse = math.sqrt(mean_squared_error(pred_df["Tuketim"].values, pred_df["sarima_pred"].values))

results = {'Prophet': [prophet_mape,prophet_rmse,prophet_r], 
        'ARIMA': [arima_mape,arima_rmse,arima_r],
        'SARIMA': [sarima_mape,sarima_rmse,sarima_r]}
results = pd.DataFrame(results,index=['MAE','RMSE','R2'])
results
```
