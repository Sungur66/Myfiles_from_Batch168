## CODING TASK #1. PLOT PIE CHART USING MATPLOTLIB

- **The plot method on Pandas Series and DataFrames is just a simple wrapper around plt.plot():**
- **Define a Pandas Dataframe with all crypto allocation in a given portfolio**
- **Note that total sum = 100%** 
- **Use matplotlib to plot a pie chart** 

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno

%matplotlib inline

import warnings

warnings.filterwarnings("ignore") 
```

```python
crypto_portfolio = pd.read_csv('crypto_daily_prices.csv')
crypto_portfolio 
```

```python
crypto_portfolio['Date'] = pd.to_datetime(crypto_portfolio['Date'])
```

```python
crypto_portfolio.set_index('Date', inplace=True)
crypto_portfolio 
```

```python
crypto_portfolio['allocations_total'] = crypto_portfolio.iloc[:, -3:-1].sum(axis=1) 

crypto_portfolio.sample(10)
```

```python
crypto_portfolio['Percentage_BTC'] = (crypto_portfolio['BTC'] / crypto_portfolio['allocations_total']) * 100 
crypto_portfolio['Percentage_ETH'] = (crypto_portfolio['ETH'] / crypto_portfolio['allocations_total']) * 100 
crypto_portfolio['Percentage_ADA'] = (crypto_portfolio['ADA'] / crypto_portfolio['allocations_total']) * 100 

crypto_portfolio.sample(5)
```

```python
crypto_portfolio_tail = crypto_portfolio.tail(1) 

crypto_allocation = crypto_portfolio_tail[['Percentage_BTC', 'Percentage_ETH', 'Percentage_ADA']]

crypto_allocation 
```

```python
crypto_allocation.rename(columns={"Percentage_BTC": "BTC", "Percentage_ETH": "ETH", "Percentage_ADA": "ADA"}, inplace = True) 
crypto_allocation 
```

```python
my_values = crypto_allocation.values.round()
my_values 
```

```python
my_labels = crypto_allocation.columns
my_labels 
```

```python
plt.figure(figsize=(8,6))

my_values = [94.931608, 5.068392, 0.002261]
plt.pie(my_values, labels=my_labels, autopct='%1.1f%%', startangle=140, explode = [0.5, 0.0, 0.0])

plt.axis('equal');
```

## CODING TASK #2. PLOT SINGLE & MULTIPLE LINE PLOTS USING MATPLOTLIB
- **Use Pandas read_csv to read crypto_daily_prices (BTC, ETH, and LTC)**
- **Use matplotlib on the Pandas DataFrame to plot the data**

```python
investments_df = pd.read_csv('crypto_daily_prices.csv')
investments_df 
```

```python
investments_df.info()
```

```python
investments_df['Date'] = pd.to_datetime(investments_df['Date'],
                                   format="%m/%d/%Y")
```

```python
investments_df.info()
```

```python
import matplotlib.dates as mdates 
```

```python
plt.figure(figsize=(12,6))

plt.plot(investments_df['Date'], investments_df['BTC'], label="BTC")

date_format = mdates.DateFormatter("%b %Y")
plt.gca().xaxis.set_major_formatter(date_format)

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('BTC Price History Over Time')
plt.legend(loc='upper left')
plt.xticks(rotation=45);  
```

```python
plt.figure(figsize=(12,6))

plt.plot(investments_df['Date'], investments_df['ETH'], label="ETH")

date_format = mdates.DateFormatter("%b %Y")
plt.gca().xaxis.set_major_formatter(date_format)

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('ETH Price History Over Time')
plt.legend(loc='upper left')
plt.xticks(rotation=45);  
```

```python
plt.figure(figsize=(12,6))

plt.plot(investments_df['Date'], investments_df['ADA'], label="ADA")

date_format = mdates.DateFormatter("%b %Y")
plt.gca().xaxis.set_major_formatter(date_format)

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('ADA Price History Over Time')
plt.legend(loc='upper left')
plt.xticks(rotation=45);  
```

**PRACTICE OPPORTUNITY #2:**
- **Use subplots attribute to plot separate figures for BTC, ETH, and ADA**

```python
fig, ax = plt.subplots(3, 1, figsize =(12, 20))

plt.suptitle("BTC & ETH & ADA Price History Over Time", fontsize = 20, y = 1)
date_format = mdates.DateFormatter("%b %Y")


ax[0].plot(investments_df['Date'], investments_df['BTC'], label="BTC")
ax[0].set_xlabel('Date')
ax[0].set_ylabel('Price')
ax[0].set_title('BTC Price History Over Time')
ax[0].xaxis.set_major_formatter(date_format)



ax[1].plot(investments_df['Date'], investments_df['ETH'], label="ETH")
ax[1].set_xlabel('Date')
ax[1].set_ylabel('Price')
ax[1].set_title('ETH Price History Over Time')
ax[1].xaxis.set_major_formatter(date_format)


ax[2].plot(investments_df['Date'], investments_df['ADA'], label="ADA")
ax[2].set_xlabel('Date')
ax[2].set_ylabel('Price')
ax[2].set_title('ADA Price History Over Time')
ax[2].xaxis.set_major_formatter(date_format)

plt.tight_layout()
plt.show()
```

## CODING TASK #4. PLOT HISTOGRAMS USING MATPLOTLIB
- **A histogram represents data using bars with various heights**
- **Each bar groups numbers into specific ranges**
- **Taller bars show that more data falls within that specific range**
- **Plot the histogram of the returns**

```python
daily_return_df = pd.read_csv('crypto_daily_returns.csv')
daily_return_df 
```

```python
mu = round(daily_return_df['BTC'].mean(), 2) # mean of distribution
sigma = round(daily_return_df['BTC'].std(), 2) # standard deviation of distribution
num_bins = 40

sns.histplot(daily_return_df['BTC'].dropna(), bins=num_bins, color = 'crimson', edgecolor='maroon', kde=True)
plt.axvline(mu,color='maroon',linestyle='-.')
plt.axvline(mu+sigma,color="maroon",linestyle='-.')
plt.axvline(mu-sigma,color="maroon",linestyle='-.')
plt.xlabel('BTC Daily Return')
plt.ylabel('Density');  
```

```python
mu = round(daily_return_df['ETH'].mean(), 2) # mean of distribution
sigma = round(daily_return_df['ETH'].std(), 2) # standard deviation of distribution
num_bins = 40

sns.histplot(daily_return_df['ETH'].dropna(), bins=num_bins, color = 'goldenrod', edgecolor='brown', kde=True)
plt.axvline(mu,color='brown',linestyle='-.')
plt.axvline(mu+sigma,color="brown",linestyle='-.')
plt.axvline(mu-sigma,color="brown",linestyle='-.')
plt.xlabel('ETH Daily Return')
plt.ylabel('Density');  
```

```python
mu = round(daily_return_df['ADA'].mean(), 2) # mean of distribution
sigma = round(daily_return_df['ADA'].std(), 2) # standard deviation of distribution
num_bins = 40

sns.histplot(daily_return_df['ADA'].dropna(), bins=num_bins, color = 'forestgreen', edgecolor='teal', kde=True)
plt.axvline(mu,color='teal',linestyle='-.')
plt.axvline(mu+sigma,color="teal",linestyle='-.')
plt.axvline(mu-sigma,color="teal",linestyle='-.')
plt.xlabel('ADA Daily Return')
plt.ylabel('Density');  
```

**PRACTICE OPPORTUNITY #4:**
- **Plot the histogram for BTC and ETH returns using 60 bins. Display all histograms on the same figure**


```python
fig, ax = plt.subplots(figsize=(10,6));
mylegend=['BTC','ETH']
ax.hist(daily_return_df['BTC'],bins=60,color='teal',edgecolor='k')
ax.hist(daily_return_df['ETH'],bins=60,color='palevioletred',edgecolor='k')
plt.xlabel('BTC & ETH Daily Return %')
plt.ylabel('Density')
plt.legend(mylegend)
plt.show() 
```

## CODING TASK #5. PLOT SCATTERPLOT & COUNTPLOT USING SEABORN
- **Seaborn is a visualization library that sits on top of matplotlib**
- **Seaborn offers enhanced features compared to matplotlib**
- **https://seaborn.pydata.org/examples/index.html*/*
- **Check out the head of the dataframe**
- **Check out the tail of the dataframe**
- **Plot scatter plot between mean area and mean smoothness**
- **Let's print out countplot to know how many samples belong to class #0 and #1**

```python
# import libraries 
import pandas as pd # Import Pandas for data manipulation using dataframes
import numpy as np # Import Numpy for data statistical analysis 
import matplotlib.pyplot as plt # Import matplotlib for data visualisation
import seaborn as sns # Statistical data visualization
```

```python
# Import Cancer data 
cancer_df = pd.read_csv('cancer.csv')
```

```python
cancer_df.head(5)
```

```python
cancer_df.info() 
```

```python
cancer_df.tail(5) 
```

```python
cancer_df['tumor'] = np.where(cancer_df['target']==1, "Malignant", "Benign")

cancer_df.sample(5)
```

```python
plt.figure(figsize=(10,5))

sns.scatterplot(data=cancer_df, x= 'mean area', y= 'mean smoothness', hue='tumor');
```

```python
sns.countplot(data=cancer_df, x='target', hue='tumor', dodge=False)

plt.xlabel('tumor', fontsize=15)
plt.ylabel('count', fontsize=15)
plt.legend(labels = ['0: Benign', '1: Malignant']); 
```

**PRACTICE OPPORTUNITY #5:**
- **Plot the scatterplot between the mean radius and mean area. Comment on the plot** 


```python
plt.figure(figsize=(10,5))

sns.scatterplot(data=cancer_df, x= 'mean radius', y= 'mean area', hue='tumor')

plt.xlabel('Mean Radius')
plt.ylabel('Mean Area')
plt.title('Scatterplot between Mean Radius and Mean Area');
```

**This scatter plot shows that there is a strong, positive and linear association between Mean Radius and Mean Area. There don't appear to be any outliers in the data.**

## CODING TASK #6. PLOT PAIRPLOT, DISPLOT, AND HEATMAPS/CORRELATIONS USING SEABORN

- **Plot the pairplot**
- **Strong correlation between the mean radius and mean perimeter, mean area and mean primeter**
- **plot the distplot** 
- **Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)**
- **KDE is used to plot the Probability Density of a continuous variable.** 

```python
sns.pairplot(cancer_df, vars=['mean radius', 'mean perimeter', 'mean area'], kind='hist', hue='tumor'); 
```

```python
plt.figure(figsize = (12,8))

corr_matrix = cancer_df[['mean area', 'mean radius', 'mean perimeter']].corr()
print(corr_matrix)

sns.heatmap(corr_matrix, annot = True, cmap='flare'); 
```

```python
sns.displot(data= cancer_df, x='mean radius', kde = True, bins=50, legend=True, hue='tumor')
plt.ylabel('Density');
```

```python
sns.displot(data= cancer_df, x='mean area', kde = True, bins=50, legend=True, hue='tumor')
plt.ylabel('Density')
plt.show() 
```

```python
sns.displot(data= cancer_df, x='mean perimeter', kde = True, bins=50, legend=True, hue='tumor')
plt.ylabel('Density');
```

**PRACTICE OPPORTUNITY #6:**
- **Plot two separate distplot for each target class #0 and target class #1**


```python
Malignant = cancer_df[cancer_df['target'] == 1]
Benign = cancer_df[cancer_df['target'] == 0] 

plt.figure(figsize=(10,6))

plt.subplot(1,2,1)
sns.distplot(Malignant['mean radius'], kde = True, bins = 30, hist_kws = dict(edgecolor="k", linewidth=2), color='red')
plt.title("Class-1 : Malignant")
plt.xlabel('Mean Radius')
plt.ylabel('Density')

plt.subplot(1,2,2)
sns.distplot(Benign['mean radius'], kde = True, bins = 30, hist_kws = dict(edgecolor="k", linewidth=2), color='blue')
plt.title("Class-0 : Benign")
plt.xlabel('Mean Radius')
plt.ylabel('Density')

plt.tight_layout(); 
```
