## CODING TASK #1. PLOT PIE CHART USING MATPLOTLIB

- **The plot method on Pandas Series and DataFrames is just a simple wrapper around plt.plot():**
- **Define a Pandas Dataframe with all crypto allocation in a given portfolio**
- **Note that total sum = 100%** 
- **Use matplotlib to plot a pie chart** 

```python
crypto_portfolio = pd.read_csv('crypto_daily_prices.csv')
crypto_portfolio 
```

```python
crypto_portfolio['Date'] = pd.to_datetime(crypto_portfolio['Date'])
```

```python
crypto_portfolio.set_index('Date', inplace=True)
crypto_portfolio.sample(3)
```

```python
crypto_portfolio['allocations_total'] = crypto_portfolio.iloc[:, -3:-1].sum(axis=1) 

crypto_portfolio.sample(10)
```

```python
crypto_portfolio['Percentage_BTC'] = (crypto_portfolio['BTC'] / crypto_portfolio['allocations_total']) * 100 
crypto_portfolio['Percentage_ETH'] = (crypto_portfolio['ETH'] / crypto_portfolio['allocations_total']) * 100 
crypto_portfolio['Percentage_ADA'] = (crypto_portfolio['ADA'] / crypto_portfolio['allocations_total']) * 100 

crypto_portfolio.sample(5)
```

```python
crypto_portfolio_tail = crypto_portfolio.tail(1) 

crypto_allocation = crypto_portfolio_tail[['Percentage_BTC', 'Percentage_ETH', 'Percentage_ADA']]

crypto_allocation 
```

```python
crypto_allocation.rename(columns={"Percentage_BTC": "BTC", "Percentage_ETH": "ETH", "Percentage_ADA": "ADA"}, inplace = True) 
crypto_allocation 
```

```python
my_values = crypto_allocation.values.round()
my_values 
```

```python
my_labels = crypto_allocation.columns
my_labels 
```

```python
plt.figure(figsize=(8,6))

my_values = [94.931608, 5.068392, 0.002261]
plt.pie(my_values, labels=my_labels, autopct='%1.1f%%', startangle=140, explode = [0.1, 0.0, 0.0])

plt.axis('equal');
```

## CODING TASK #4. PLOT HISTOGRAMS USING MATPLOTLIB
- **A histogram represents data using bars with various heights**
- **Each bar groups numbers into specific ranges**
- **Taller bars show that more data falls within that specific range**
- **Plot the histogram of the returns**

```python
daily_return_df = pd.read_csv('crypto_daily_returns.csv')
daily_return_df
```

```python
mu = round(daily_return_df['BTC'].mean(), 2) # mean of distribution
sigma = round(daily_return_df['BTC'].std(), 2) # standard deviation of distribution
 
num_bins = 40
```

```python
mu
```

```python
sigma
```

```python
mylegend=[mu,sigma]
plt.hist(daily_return_df['BTC'],num_bins,color='c',edgecolor='k')
plt.axvline(mu,color='b',linestyle='-.')
plt.axvline(mu+sigma,color="r",linestyle='dashed')
plt.axvline(mu-sigma,color="y",linestyle='dashed')
plt.xlabel('Daily Return %')
plt.ylabel('Frequency')
plt.legend(mylegend)
plt.show()
```

**PRACTICE OPPORTUNITY #4:**
- **Plot the histogram for BTC and ETH returns using 60 bins. Display all histograms on the same figure**


```python
fig, ax = plt.subplots(figsize=(12,8));
mylegend=['BTC','ETH']
ax.hist(daily_return_df['BTC'],bins=60,color='orange',edgecolor='k')
ax.hist(daily_return_df['ETH'],bins=60,color='skyblue',edgecolor='k')
plt.xlabel('Daily Return %')
plt.ylabel('Frequency')
plt.legend(mylegend)
plt.show()
```

Using Matplotlib, plot a pie chart that shows these allocations. Use 'explode’ attribute to increase the separation between AAPL and the rest of the portfolio.
    - Using Matplotlib, plot the histogram for United Airlines and P&G returns using 40 bins with red color. Display the mean and Standard deviation for both stocks on top of the figure. What did you conclude from this graph?

```python
# Import both datasets using Pandas.

stock_price = pd.read_csv("stocks_daily_prices.csv")
stock_return = pd.read_csv("stocks_daily_returns.csv")
stock_price.sample(5)
```

```python
stock_return.sample(5)
```

```python
# Using Matplotlib, plot lineplots that display all 4 stocks daily prices on one single figure.

plt.figure(figsize=(10,6))
plt.plot(stock_price["Date"], stock_price["AAPL"], "b", label = "AAPL")
plt.plot(stock_price["Date"], stock_price["JPM"], "r", label = "JPM")
plt.plot(stock_price["Date"], stock_price["PG"], "y", label = "PG")
plt.plot(stock_price["Date"], stock_price["UAL"], "g", label = "UAL")

plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Price by Date")

plt.legend();
```

```python
# Using Matplotlib, plot 4 stocks daily prices on multiple subplots.

fig, ax = plt.subplots(2, 2, figsize = (18, 10))

ax[0,0].plot(stock_price["Date"], stock_price["AAPL"], "b")
ax[0,0].set_xlabel("AAPL")

ax[0,1].plot(stock_price["Date"], stock_price["JPM"], "r")
ax[0,1].set_xlabel("JPM")

ax[1,0].plot(stock_price["Date"], stock_price["PG"], "y")
ax[1,0].set_xlabel("PG")

ax[1,1].plot(stock_price["Date"], stock_price["UAL"], "g")
ax[1,1].set_xlabel("UAL")

plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Price by Date")

plt.legend();
```

```python
# Using Matplotlib, plot the 4 plots on subplots next to each other (all figures in one row).

fig, ax = plt.subplots(1, 4, figsize = (20, 5))

ax[0].plot(stock_price["Date"], stock_price["AAPL"], "b")
ax[0].set_xlabel("AAPL")

ax[1].plot(stock_price["Date"], stock_price["JPM"], "r")
ax[1].set_xlabel("JPM")

ax[2].plot(stock_price["Date"], stock_price["PG"], "y")
ax[2].set_xlabel("PG")

ax[3].plot(stock_price["Date"], stock_price["UAL"], "g")
ax[3].set_xlabel("UAL")

plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Price by Date")

plt.legend();
```

```python
# Using Matplotlib, plot the scatterplot between Apple and JP Morgan daily returns.

plt.figure(figsize = (20, 12))
labels=["Apple","JP Morgan"]
plt.scatter(stock_return['Date'], stock_return['AAPL'])
plt.scatter(stock_return['Date'], stock_return['JPM'])
plt.title("Daily Returns of Apple und JP Morge")
plt.xlabel("Date")
plt.ylabel('Daily Returns')
plt.xticks([])
plt.legend(labels)
plt.show();
```

```python
# Using Seaborn, plot similar scatterplot between Apple and JP Morgan daily returns.
plt.figure(figsize=(9,5))
sns.scatterplot(data=stock_return, x = "Date", y = "AAPL")
sns.scatterplot(data=stock_return, x = "Date", y = "JPM");
```

```python
# Assume that you decided to become bullish on AAPL and you allocated 70% of your assets in it. 
# You also decided to equally divide the rest of your assets in other stocks (JPM, PG, and UAL). 
# Using Matplotlib, plot a pie chart that shows these allocations. 
# Use 'explode’ attribute to increase the separation between AAPL and the rest of the portfolio.

stocks_portfolio={'Stocks':['Apple','JP Morgan Chase','Procter and Gamble','United Airlines'],
                  'Ticker':["AAPL","JPM","PG","UAL"], 
                  'Allocation':[stock_price["AAPL"].sum(),
                                stock_price["JPM"].sum(),
                                stock_price["PG"].sum(),
                                stock_price["UAL"].sum()]}
stocks_allocation= pd.DataFrame(stocks_portfolio)
stocks_allocation
```

```python
plt.figure(figsize=(8,5))
plt.pie(x=stocks_allocation["Allocation"],autopct="%.2f%%",explode=(0,0,0.1,0),labels=stocks_allocation['Ticker'])
plt.show()
```

```python
stocks_allocation.iloc[0,2]=70
stocks_allocation.iloc[1:4,2]=10
stocks_allocation
```

```python
plt.figure(figsize=(8,5))
plt.pie(x=stocks_allocation["Allocation"],autopct="%.2f%%",explode=(0.1,0,0,0),labels=stocks_allocation['Ticker'])
plt.show()
```

```python
# Using Matplotlib, plot the histogram for United Airlines and P&G returns using 40 bins with red color. Display the mean and Standard deviation for both stocks on top of the figure. 

mean_UAL = round(stock_price["UAL"].mean(), 2)
mean_PG = round(stock_price["PG"].mean(), 2)
std_UAL = round(stock_price["UAL"].std(), 2)
std_PG = round(stock_price["PG"].std(), 2)
mylegend1 = ["mean_UAL", mean_UAL, "std_UAL", std_UAL]
mylegend2 = ["mean_PG", mean_PG, "std_PG", std_PG]

plt.hist([stock_price["UAL"], stock_price["PG"]], 
         bins= 40,color=['r', 'b'], 
         label=["UAL", "PG"])

plt.xlabel('Count')
plt.ylabel('Price')
plt.legend([mylegend1, mylegend2]);
```

# What did you conclude from this graph?

##### UAL's price range and value are lower than PG, PG's price values are higher and price diversity is greater.

#### The correlation value between UAL and PG is approximately 0.3, so we cannot talk about a correlation between them.
