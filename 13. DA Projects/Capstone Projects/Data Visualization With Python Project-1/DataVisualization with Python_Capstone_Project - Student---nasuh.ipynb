## weather_code, is_holiday,is_weekend and season will not be left as float

```python
"""season - category field meteorological seasons: 0-spring ; 1-summer; 2-fall; 3-winter."""

df['season'].replace(0.0,"Spring",inplace=True)
df['season'].replace(1.0,"Summer",inplace=True)
df['season'].replace(2.0,"Fall",inplace=True)
df['season'].replace(3.0,"Winter",inplace=True)
```

```python
"""is_holiday - boolean field - 1 holiday / 0 non holiday"""

df['is_holiday'].replace(1.0,'Holiday',inplace=True)
df['is_holiday'].replace(0.0,'Non-Holiday',inplace=True)
```

```python
"""is_weekend - boolean field - 1 if the day is weekend"""

df['is_weekend'].replace(1.0, 'Weekend', inplace=True)
df['is_weekend'].replace(0.0, 'Weekday', inplace=True)
```

```python
"""weather_code => 1:'Clear', 2:'Few_Scattered_Clouds', 3:'Broken_clouds',
4:'Cloudy', 7:'Rain_Light_Rain_Shower', 10:'Rain_with_Thunderstorm',
26:'Snowfall', 94:'Freezing_Fog'"""

df['weather_code'].replace(1, 'Clear', inplace=True)
df['weather_code'].replace(2, 'Few_Scattered_Clouds', inplace=True)
df['weather_code'].replace(3, 'Broken_clouds', inplace=True)
df['weather_code'].replace(4, 'Cloudy', inplace=True)
df['weather_code'].replace(7, 'Rain_Light_Rain_Shower', inplace=True)
df['weather_code'].replace(10, 'Rain_with_Thunderstorm', inplace=True)
df['weather_code'].replace(26, 'Snowfall', inplace=True)
df['weather_code'].replace(94, 'Freezing_Fog', inplace=True)
```

```python
df.sample(5)
```

## we have converted float type values to object type values for categorical fileds so that we can visualize them properly

## -------Bike Demand/Share by Season--------

```python
Bike_Share_by_Season = df.groupby('season')['cnt'].sum()
Bike_Share_by_Season
```

## ----Let's change values to single digit values (in millions)----

```python
Bike_Share_by_Season = (df.groupby('season')['cnt'].sum()/1000).astype('int64')
Bike_Share_by_Season
```

```python
fig, ax = plt.subplots(1, 2 , figsize = (16, 6))
fig.suptitle("Bike Share by Season",ha="center")

Bike_Share_by_Season.plot(kind='pie',autopct="%.2f%%",explode=(0,0.2,0,0,),labels=Bike_Share_by_Season.index,ax=ax[0]);
ax=sns.barplot(data=df, x=Bike_Share_by_Season.index, y=Bike_Share_by_Season.values,ax = ax[1])
for i in ax.containers:
    ax.bar_label(i)

plt.ylabel("Count (in thousands)")
plt.show()
```

## Bike sharing is the least during the winter season

## ----Bike Share on Holidays/Non-Holidays----

```python
Holidays_Bike_Share  = (df.groupby("is_holiday")['cnt'].sum()/1000).astype("int64")
Holidays_Bike_Share
```

```python
fig, ax = plt.subplots();
ax.bar(Holidays_Bike_Share.index,Holidays_Bike_Share.values)
ax.set_title("Bike Share on Holidays/Non-Holidays ")
ax.set_ylabel('Count (in thousands)');
for i in ax.containers:
    ax.bar_label(i);
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))
ax[0] = sns.countplot(data=df, x="is_holiday", ax=ax[0])
ax[1] = sns.countplot(data=df, x="is_weekend", ax=ax[1])
ax[0].set_title("Bike Share on Holidays/Non-Holidays ");
ax[1].set_title("Bike Share on Weekdays/Weekends");
```

## *----Bike Shares by Weekends/Weekdays----*

```python
bike_share_by_weekend  = (df.groupby('is_weekend')['cnt'].sum()/1000).astype('int64')
bike_share_by_weekend
```

```python
fig, ax = plt.subplots()
ax.bar(bike_share_by_weekend.index,bike_share_by_weekend.values)
ax.set_title("Weekend Bike Shares ")
ax.set_ylabel('Count (in thousands)')
for i in ax.containers:
    ax.bar_label(i);
```

## This graph along with the bike shares by holidays suggest that majority of people use bike sharing as part of their daily routine in London. Bike share is used around businesses not residential areas. Weekends are much busier than holidays as a second note here.

```python
fig, ax = plt.subplots(2,2, figsize=(12, 12))

# first plot by season
sns.barplot(data=df, x="season", y="cnt", ax = ax[0][0], ci=0)
ax[0][0].set_xlabel("Season")
ax[0][0].set_ylabel('Count')
ax[0][0].set_title('Bike Share by Season')
##ax[0][0].set_xlabel('') # x-label doesn't show by doing this

# second plot by holiday
sns.countplot(data=df, x="is_holiday", ax=ax[0][1])
ax[0][1].set_xlabel("Holidays/Non-holidays")
ax[0][1].set_ylabel('Count')
ax[0][1].set_title('Bike Share by Holidays')
##ax[0][1].set_xlabel('') # x-label doesn't show by doing this

# third plot by weekend
sns.countplot(data=df, x="is_weekend", ax=ax[1][0])
ax[1][0].set_xlabel("Weekday/Weekend")
ax[1][0].set_ylabel('Count')
ax[1][0].set_title('Bike Share by Weekends')
##ax[1][0].set_xlabel('') # x-label doesn't show by doing this

# fourth plot by weather code
sns.countplot(data=df, x="weather_code", ax=ax[1][1])
ax[1][1].set_xlabel("Weather Conditions")
ax[1][1].set_ylabel('Count')
ax[1][1].set_title('Bike Share by Weather Conditions')
ax[1][1].set_xticklabels(ax[1][1].get_xticklabels(), rotation=90, ha='right')
##ax[1][1].set_xlabel('') # x-label doesn't show by doing this

fig.tight_layout()
plt.show();
```

```python
df_discrete = ["season", "is_holiday", "is_weekend", "weather_code"]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,8))
fig.tight_layout()

for i, ax in enumerate(axes.flat):
    feature = df_discrete[i]
    
    # Create a boxplot for each feature
    box_plot=sns.boxplot(data=df, y=feature,x ="cnt",ax=ax)

    ax.set_xlabel("Count")
    ax.set_ylabel(feature)
    ax.set_title('Distribution of ' + feature)
    
fig.tight_layout()
plt.show()
```
