## <span style=" color:red">Determines 
    
Features
    
- timestamp - timestamp field for grouping the data
- cnt - the count of a new bike shares
- t1 - real temperature in C
- t2 - temperature in C “feels like”
- hum - humidity in percentage
- wind_speed - wind speed in km/h
- weather_code - category of the weather
- is_holiday - boolean field - 1 holiday / 0 non holiday
- is_weekend - boolean field - 1 if the day is weekend
- season - category field meteorological seasons: 0-spring ; 1-summer; 2-fall; 3-winter.
    
"weather_code" category description:

- 1 = Clear ; mostly clear but have some values with haze/fog/patches of fog/ fog in vicinity
- 2 = scattered clouds / few clouds
- 3 = Broken clouds
- 4 = Cloudy
- 7 = Rain/ light Rain shower/ Light rain
- 10 = rain with thunderstorm
- 26 = snowfall
- 94 = Freezing Fog
    
Initially, the task of discovering data will be waiting for you as always. Recognize features, detect missing values, outliers etc. Review the data from various angles in different time breakdowns. For example, visualize the distribution of bike shares by day of the week. With this graph, you will be able to easily observe and make inferences how people's behavior changes daily. Likewise, you can make hourly, monthly, seasonally etc. analyzes. In addition, you can analyze correlation of variables with a heatmap.

<span style=" color:red">1. Import Libraries
</span>

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno

%matplotlib inline

import warnings

warnings.filterwarnings("ignore") 
```

<div class="alert alert-warning alert-info">
<span style=" color:red">2. Read Dataset
</span>

```python
bike = pd.read_csv("store_sharing.csv", encoding='windows-1252')
bike 
```

```python
bike.info() 
```

```python
bike.describe().T 
```

<div class="alert alert-warning alert-info">
<span style=" color:red">3. Check missing values and if there are any dublicate rows or not.
</span>

```python
bike.isnull().sum() 
```

```python
msno.bar(bike);  
```

```python
bike.duplicated().sum() 
```

```python
bike['is_holiday_2'] = np.where(bike['is_holiday'] == 0, 'non holiday',
                                'holiday')
bike['is_weekend_2'] = np.where(bike['is_weekend'] == 0, 'weekday', 'weekend')
bike 
```

```python
bike['weather_code'].unique() 
```

```python
def weather_code_2(value):
    if value == 1:
        return "Clear"
    elif value == 2:
        return "Scattered Clouds"
    elif value == 3:
        return "Broken Clouds"
    elif value == 4:
        return "Cloudy"
    elif value == 7:
        return "Rain"
    elif value == 10:
        return "Rain with Thunderstorm"
    elif value == 26:
        return "Snow Fall"

bike['weather_code_2'] = bike['weather_code'].map(weather_code_2)

bike.head() 
```

```python
bike['season'].unique() 
```

```python
bike['is_holiday_2'].value_counts() 
```

```python
def season_2(value):
    if value == 0:
        return "Spring"
    elif value == 1:
        return "Summer"
    elif value == 2:
        return "Fall"
    elif value == 3:
        return "Winter"


bike['season_2'] = bike['season'].map(season_2)

bike.head() 
```

```python
bike.isnull().sum() 
```

```python
bike.info() 
```

```python
bike['is_holiday_2'] = bike['is_holiday_2'].astype('category')
bike['is_weekend_2'] = bike['is_weekend_2'].astype('category')
bike['weather_code_2'] = bike['weather_code_2'].astype('category')
bike['season_2'] = bike['season_2'].astype('category')

bike.info() 
```

<div class="alert alert-warning alert-info">
<span style=" color:red">4. Plot the distribution of various discrete features on (Season, haliday, weekend and weathercode)
</span>


```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

ax[0] = sns.barplot(data=bike, x="is_holiday_2",y = "cnt", ax=ax[0])
ax[1] = sns.barplot(data=bike, x="is_weekend_2",y= "cnt", ax=ax[1]); 
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

ax[0] = sns.barplot(data=bike, x="weather_code_2",y="cnt", ax=ax[0])
ax[1] = sns.barplot(data=bike, x="season_2", y="cnt", ax=ax[1])

plt.setp(ax[0].get_xticklabels(),
         rotation=45,
         ha="right",
         rotation_mode="anchor"); 
```

<span style=" color:red">9. Plot bike shares over time use lineplot.
</span>

```python
plt.figure(figsize=(10, 6))
plt.plot(bike.index, bike['cnt'], marker='o', linestyle='-')

# Set labels and title
plt.xlabel('Date')
plt.ylabel('Bike Shares')
plt.title('Bike Shares Over Time')

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45)

plt.grid(True)  # Add grid lines (optional)

plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(data=bike, x='Year', y="cnt")
plt.title("Bike Share Over The Years")
plt.xlabel("Year")
plt.xticks(bike['Year'].unique().astype(int))


plt.show()
```

# <span style=" color:red">Conclusions
    

    

</span>

In this work, I studied bike-sharing data in London to get some insights into people's preferences about bicycle use according to the days of the week, weather conditions, and seasons. 

In the investigation parts of the study, I got some information about the data for exploratory research. And I added categorical columns to it to get more clarity in the graphs. 

In the visualization parts of the study, I used distribution graphs to understand the behaviors of people who live in London. I observed that bike-sharing counts significantly increased during the weekdays, typically before and after work shifts. 
With this sight, we can claim that people use bike-sharing as transportation to go to work and return. 

The weather conditions can affect people's preferences for transportation choices. The amount of bike-sharing on wet days is less than on clear days. I got the same results when analyzing the season effects, and I observed the lowest bike-sharing rates are typically in the winter seasons. 

The results of this study show that bike sharing helps increase bicycle use in the city, and we can regard it as a possible decrease in traffic congestion.
